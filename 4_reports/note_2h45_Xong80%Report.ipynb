### **1. Tính cấp thiết của đề tài khóa luận**
Ngành công nghệ thông tin (CNTT) đang đóng vai trò quan trọng trong quá trình chuyển đổi số và sự phát triển kinh tế toàn cầu. Tại Việt Nam, sự tăng trưởng mạnh mẽ của các công ty CNTT không chỉ góp phần thúc đẩy nền kinh tế mà còn thu hút sự quan tâm lớn từ các nhà đầu tư trên thị trường chứng khoán. Tuy nhiên, sự biến động phức tạp của giá cổ phiếu ngành CNTT, chịu ảnh hưởng từ các xu hướng công nghệ, yếu tố kinh tế vĩ mô và tâm lý thị trường, đã đặt ra thách thức lớn trong việc dự đoán chính xác giá cổ phiếu.

Dự đoán giá cổ phiếu là một bài toán chuỗi thời gian phức tạp, yêu cầu các công cụ và mô hình phân tích tiên tiến để hỗ trợ nhà đầu tư trong việc tối ưu hóa lợi nhuận và giảm thiểu rủi ro. Các mô hình như:
- **ARIMA**: Hiệu quả trong dự báo ngắn hạn và xử lý chuỗi thời gian tuyến tính.
- **LSTM**: Phù hợp với dữ liệu phi tuyến tính và có khả năng nắm bắt mối quan hệ dài hạn trong dữ liệu.
- **Ridge Linear Regression**: Kết hợp các đặc trưng tài chính và điều kiện ràng buộc, mang lại tính ổn định trong phân tích chuỗi thời gian.

Với tiềm năng ứng dụng mạnh mẽ của trí tuệ nhân tạo và học máy trong dự báo tài chính, đề tài "Ứng dụng Ridge Linear Regression, ARIMA và LSTM trong dự đoán giá cổ phiếu ngành công nghệ thông tin" được lựa chọn để nghiên cứu và triển khai. Đề tài không chỉ hỗ trợ nhà đầu tư trong việc đưa ra quyết định chiến lược mà còn mở rộng nghiên cứu ứng dụng trí tuệ nhân tạo trong lĩnh vực tài chính tại Việt Nam.

- Phân tích các yếu tố ảnh hưởng đến độ chính xác khi dự đoán giá cổ phiếu ngành CNTT, bao gồm xu hướng thị trường, tính mùa vụ và các chỉ báo kỹ thuật.

#### **Thu thập và xử lý dữ liệu từ thị trường chứng khoán Việt Nam**

#### 3.1.3. Chất Lượng Dữ Liệu

Đảm bảo chất lượng dữ liệu là yếu tố then chốt để xây dựng các mô hình dự báo chính xác. Các bước kiểm tra chất lượng dữ liệu bao gồm:

- **Kiểm Tra Giá Trị Thiếu (Missing Values):** Đảm bảo rằng không có giá trị thiếu trong các cột quan trọng như giá mở cửa, giá đóng cửa và khối lượng giao dịch. Nếu có, tiến hành các bước xử lý như điền giá trị trung bình hoặc loại bỏ các dòng dữ liệu bị thiếu.

- **Kiểm Tra Ngoại Lai (Outliers):** Xác định và xử lý các giá trị ngoại lai có thể ảnh hưởng đến độ chính xác của mô hình. Sử dụng các phương pháp thống kê như IQR (Interquartile Range) để phát hiện và xử lý ngoại lai.

- **Kiểm Tra Tính Đồng Nhất (Consistency):** Đảm bảo rằng các dữ liệu được thu thập từ các nguồn khác nhau không có sự mâu thuẫn về thông tin. Sử dụng các công cụ so sánh dữ liệu để kiểm tra tính đồng nhất.

- **Kiểm Tra Độ Tin Cậy (Reliability):** Xác minh rằng dữ liệu thu thập được là chính xác và đáng tin cậy, không bị sai lệch hoặc lỗi từ quá trình thu thập.

### 3.2. Tiền Xử Lý Dữ Liệu

Sau khi thu thập dữ liệu, bước tiếp theo là tiền xử lý để đảm bảo rằng dữ liệu sẵn sàng cho quá trình phân tích và xây dựng mô hình. Các bước tiền xử lý dữ liệu bao gồm:

#### 3.2.1. Làm Sạch Dữ Liệu (Data Cleaning)

- **Xử Lý Giá Trị Thiếu:**
  - **Kiểm Tra Giá Trị Thiếu:**
    ```python
    import pandas as pd

    # Đọc dữ liệu cổ phiếu FPT
    fpt_df = pd.read_csv('data/raw20192024/FPT_stock_data.csv')

    # Kiểm tra giá trị thiếu
    missing_values = fpt_df.isnull().sum()
    print(missing_values)
    ```
  - **Điền Giá Trị Thiếu:**
    ```python
    # Điền giá trị thiếu bằng phương pháp trung bình
    fpt_df['close'].fillna(fpt_df['close'].mean(), inplace=True)
    ```

- **Xóa Các Dòng Bị Trùng Lặp:**
  ```python
  # Xóa các dòng trùng lặp
  fpt_df.drop_duplicates(inplace=True)
  ```

- **Xử Lý Ngoại Lai:**
  ```python
  # Sử dụng IQR để phát hiện ngoại lai
  Q1 = fpt_df['close'].quantile(0.25)
  Q3 = fpt_df['close'].quantile(0.75)
  IQR = Q3 - Q1

  # Xác định giới hạn ngoại lai
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Loại bỏ ngoại lai
  fpt_df = fpt_df[(fpt_df['close'] >= lower_bound) & (fpt_df['close'] <= upper_bound)]
  ```

#### 3.2.2. Chuyển Đổi và Chuẩn Hóa Dữ Liệu

- **Chuẩn Hóa Dữ Liệu (Normalization):**
  - **Mục đích:** Đưa các giá trị về cùng một khoảng để mô hình học máy có thể xử lý hiệu quả hơn.
  - **Sử dụng Min-Max Scaler:**
    ```python
    from sklearn.preprocessing import MinMaxScaler

    scaler = MinMaxScaler(feature_range=(0, 1))
    fpt_df[['close', 'volume']] = scaler.fit_transform(fpt_df[['close', 'volume']])
    ```

- **Biến Đổi Dữ Liệu (Data Transformation):**
  - **Tính Tỷ Suất Sinh Lời Hàng Ngày (Daily Returns):**
    ```python
    fpt_df['daily_returns'] = fpt_df['close'].pct_change()
    ```

#### 3.2.3. Biến Đổi Dữ Liệu (Data Transformation)

    ```python
    # Tính MA20 và MA50
    ```

  - **Không Shuffle Dữ Liệu Do Bản Chất Time Series:**

    ```python
    from sklearn.model_selection import train_test_split
    ```

#### 3.2.5. Xử Lý Dữ Liệu Không Cân Bằng (Imbalanced Data)

Trong bài toán dự đoán giá cổ phiếu (hồi quy), vấn đề dữ liệu không cân bằng thường không được đặt lên hàng đầu như trong các bài toán phân loại. Tuy nhiên, nếu bài toán chuyển sang phân loại (ví dụ: dự đoán giá tăng hay giảm), thì việc xử lý dữ liệu không cân bằng trở nên quan trọng hơn.

- **Phương Pháp Xử Lý:**
  - **Over-sampling:** Tăng số lượng mẫu trong lớp thiểu số.
  - **Under-sampling:** Giảm số lượng mẫu trong lớp đa số.
  - **Sử dụng Các Thuật Toán Phù Hợp:** Chọn các thuật toán không bị ảnh hưởng nhiều bởi dữ liệu không cân bằng.

- **Ví Dụ:**
  ```python
  from imblearn.over_sampling import SMOTE

  smote = SMOTE(random_state=42)
  ```

### 4.1. Trích Xuất Đặc Trưng (Feature Extraction)

Trích xuất đặc trưng là quá trình biến đổi dữ liệu thô thành các biến có ý nghĩa và hữu ích cho việc xây dựng mô hình học máy. Dưới đây là các loại đặc trưng được sử dụng trong nghiên cứu này:

#### 4.1.1. **Basic Price Features**

Các đặc trưng cơ bản liên quan đến giá cổ phiếu, giúp phản ánh tình hình thị trường hàng ngày.

- **Daily Price Range:**
  
  \[
  \text{Range} = \text{High} - \text{Low}
  \]
  
  *Giải thích:* Khoảng giá trong ngày đo lường sự biến động của cổ phiếu. Giá Range lớn cho thấy cổ phiếu có sự biến động mạnh trong ngày đó.

  ```python
  df['Range'] = df['High'] - df['Low']
  ```

- **Price Change:**
  
  \[
  \text{Price Change} = \text{Close} - \text{Open}
  \]
  
  *Giải thích:* Thay đổi giá trong ngày phản ánh sự biến động từ mở cửa đến đóng cửa. Giá tăng dương cho thấy sự tăng giá, ngược lại.

  ```python
  df['Price_Change'] = df['Close'] - df['Open']
  ```

- **Percentage Change:**
  
  \[
  \text{Percentage Change} = \left( \frac{\text{Close} - \text{Open}}{\text{Open}} \right) \times 100
  \]
  
  *Giải thích:* Tỷ lệ thay đổi giá giúp chuẩn hóa sự biến động cho các cổ phiếu có mức giá khác nhau.

  ```python
  df['Percentage_Change'] = ((df['Close'] - df['Open']) / df['Open']) * 100
  ```

- **Average Price:**
  
  \[
  \text{Average Price} = \frac{\text{Open} + \text{High} + \text{Low} + \text{Close}}{4}
  \]
  
  *Giải thích:* Giá trung bình trong ngày cung cấp một ước lượng về mức giá trung bình của cổ phiếu.

  ```python
  df['Average_Price'] = (df['Open'] + df['High'] + df['Low'] + df['Close']) / 4
  ```

- **Relative Price:**
  
  \[
  \text{Relative Price} = \frac{\text{Close} - \text{Close}_{n \text{days ago}}}{\text{Close}_{n \text{days ago}}}
  \]
  
  *Giải thích:* So sánh giá đóng cửa hiện tại với giá đóng cửa của các ngày trước giúp đánh giá xu hướng giá.

  ```python
  n = 5
  df[f'Relative_Price_{n}'] = (df['Close'] - df['Close'].shift(n)) / df['Close'].shift(n)
  ```

#### 4.1.2. **Rolling Window Statistics**

Các đặc trưng thống kê dựa trên cửa sổ trượt giúp nắm bắt xu hướng và biến động theo thời gian.

- **Moving Averages (Đường Trung Bình Động):**

  - **Simple Moving Average (SMA):**
    
    \[
    \text{SMA}_n = \frac{1}{n} \sum_{i=0}^{n-1} \text{Close}_{t-i}
    \]
    
    *Giải thích:* SMA trung bình giá đóng cửa trong n ngày giúp xác định xu hướng giá.

    ```python
    n = 20
    df[f'SMA_{n}'] = df['Close'].rolling(window=n).mean()
    ```

  - **Exponential Moving Average (EMA):**
    
    *Giải thích:* EMA đặt trọng số lớn hơn vào các giá gần đây, nhạy cảm hơn với biến động giá mới.

    ```python
    n = 20
    df[f'EMA_{n}'] = df['Close'].ewm(span=n, adjust=False).mean()
    ```

- **Rolling Standard Deviation:**
  
  \[
  \text{Rolling Std} = \text{Std}(\text{Close over n days})
  \]
  
  *Giải thích:* Đo lường độ biến động của giá trong một khoảng thời gian nhất định.

  ```python
  n = 20
  df[f'Rolling_Std_{n}'] = df['Close'].rolling(window=n).std()
  ```

- **Bollinger Bands:**
  
  \[
  \text{Upper Band} = \text{SMA}_n + (2 \times \text{Rolling Std}_n)
  \]
  \[
  \text{Lower Band} = \text{SMA}_n - (2 \times \text{Rolling Std}_n)
  \]
  
  *Giải thích:* Các dải Bollinger cung cấp mức hỗ trợ và kháng cự, giúp xác định khi cổ phiếu đang quá mua hoặc quá bán.

  ```python
  n = 20
  df[f'Upper_Band_{n}'] = df[f'SMA_{n}'] + (2 * df[f'Rolling_Std_{n}'])
  df[f'Lower_Band_{n}'] = df[f'SMA_{n}'] - (2 * df[f'Rolling_Std_{n}'])
  ```

#### 4.1.3. **Momentum Indicators**

Các chỉ số động lượng giúp đánh giá sức mạnh và hướng đi của xu hướng giá.

- **Relative Strength Index (RSI):**
  
  *Giải thích:* RSI đo lường tốc độ và thay đổi của chuyển động giá, thường sử dụng trong khoảng từ 0 đến 100 để xác định tình trạng quá mua (>70) hoặc quá bán (<30).

  ```python
  import ta
  
  df['RSI_14'] = ta.momentum.RSIIndicator(close=df['Close'], window=14).rsi()
  ```

- **Moving Average Convergence Divergence (MACD):**
  
  *Giải thích:* MACD kết hợp hai EMA để thể hiện mối quan hệ giữa chúng, giúp phát hiện sự thay đổi xu hướng.

  ```python
  ema_12 = df['Close'].ewm(span=12, adjust=False).mean()
  ema_26 = df['Close'].ewm(span=26, adjust=False).mean()
  df['MACD'] = ema_12 - ema_26
  df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
  ```

- **On-Balance Volume (OBV):**
  
  *Giải thích:* OBV đo lường áp lực mua và bán bằng cách cộng thêm khối lượng vào OBV khi giá tăng và trừ khối lượng khi giá giảm.

  ```python
  df['OBV'] = ta.volume.OnBalanceVolumeIndicator(close=df['Close'], volume=df['Volume']).on_balance_volume()
  ```

#### 4.1.4. **Volatility Features**

Các đặc trưng đo lường mức độ biến động của cổ phiếu, giúp đánh giá rủi ro.

- **Daily Volatility:**
  
  \[
  \text{Daily Volatility} = \frac{\text{High} - \text{Low}}{\text{Open}}
  \]
  
  *Giải thích:* Tỷ lệ biến động hàng ngày giúp đánh giá mức độ dao động của cổ phiếu so với giá mở cửa.

  ```python
  df['Daily_Volatility'] = (df['High'] - df['Low']) / df['Open']
  ```

- **Historical Volatility:**
  
  \[
  \text{Historical Volatility} = \text{Std}(\text{Returns over n days})
  \]
  
  *Giải thích:* Độ lệch chuẩn của tỷ suất sinh lời trong một khoảng thời gian nhất định, đo lường sự biến động lịch sử.

  ```python
  n = 30
  df['Historical_Volatility'] = df['Daily_Returns'].rolling(window=n).std() * np.sqrt(252)  # Annualized
  ```

- **Average True Range (ATR):**
  
  *Giải thích:* ATR đo lường sự biến động bằng cách tính toán khoảng cách thực sự trung bình giữa các ngày giao dịch.

  ```python
  df['ATR_14'] = ta.volatility.AverageTrueRange(high=df['High'], low=df['Low'], close=df['Close'], window=14).average_true_range()
  ```

#### 4.1.5. **Technical Indicators**

Các chỉ số kỹ thuật hỗ trợ việc dự đoán xu hướng và động lượng giá.

- **Stochastic Oscillator:**
  
  *Giải thích:* So sánh giá đóng cửa hiện tại với phạm vi giá trong một khoảng thời gian nhất định để xác định tình trạng quá mua hoặc quá bán.

  ```python
  df['Stochastic_Oscillator'] = ta.momentum.StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=14).stoch()
  ```

- **Williams %R:**
  
  *Giải thích:* Chỉ số đo lường mức độ quá mua hoặc quá bán, tương tự như Stochastic Oscillator nhưng có thang điểm khác (-100 đến 0).

  ```python
  df['Williams_%R'] = ta.momentum.WilliamsRIndicator(high=df['High'], low=df['Low'], close=df['Close'], window=14).williams_r()
  ```

#### 4.1.6. **Trend Features**

Các đặc trưng liên quan đến xu hướng giá và khối lượng giao dịch.

- **Price Trend:**
  
  *Giải thích:* Xác định xu hướng giá dựa trên độ dốc của giá trong một khoảng thời gian nhất định.

  ```python
  n = 10
  ```

Các đặc trưng dựa trên giá trị quá khứ, giúp mô hình học từ dữ liệu lịch sử.

- **Lagged Prices:**

Các đặc trưng liên quan đến tỷ lệ và tương quan giữa các yếu tố trong dữ liệu.

- **Price-to-Volume Ratio:**
  
  *Giải thích:* Tỷ lệ giữa giá đóng cửa và khối lượng giao dịch, giúp hiểu mức độ giao dịch tại các mức giá khác nhau.

  ```python
  df['Price_to_Volume'] = df['Close'] / df['Volume']
  ```

  ```python
  companies = ['FPT', 'CMG', '...']  # Danh sách các công ty
  ```

Các đặc trưng dựa trên thời gian, giúp mô hình nhận biết các mô hình theo mùa hoặc theo ngày trong tuần.

## CHƯƠNG 5: TRIỂN KHAI MÔ HÌNH

Chương này trình bày chi tiết về việc **triển khai các mô hình dự báo giá cổ phiếu** sử dụng các kỹ thuật học máy và thống kê. ### 5.1. Mô hình ARIMA

#### 5.1.1. Giới Thiệu về ARIMA

**ARIMA** (AutoRegressive Integrated Moving Average) là một trong những mô hình thống kê phổ biến nhất để dự báo chuỗi thời gian. ARIMA kết hợp ba thành phần chính:

- **AR (AutoRegressive):** Phần tự hồi quy, dựa trên mối quan hệ tuyến tính giữa giá trị hiện tại và các giá trị trước đó.
- **I (Integrated):** Phần tích hợp, liên quan đến việc lấy sai phân để làm cho chuỗi thời gian trở nên ổn định.
- **MA (Moving Average):** Phần trung bình động, dựa trên mối quan hệ giữa giá trị hiện tại và các sai số trước đó.

#### 5.1.2. Áp Dụng Mô Hình ARIMA

**Bước 1: Xử Lý Tính Dừng (Stationarity)**

Chuỗi thời gian cần phải ổn định để ARIMA hoạt động hiệu quả. Một chuỗi thời gian ổn định có nghĩa là các thống kê cơ bản như trung bình và phương sai không thay đổi theo thời gian.

- **Kiểm Tra Tính Dừng:** Sử dụng kiểm định Augmented Dickey-Fuller (ADF) để xác định tính dừng của chuỗi thời gian.

  ```python
  from statsmodels.tsa.stattools import adfuller

  result = adfuller(df['Close'])
  print('ADF Statistic:', result[0])
  print('p-value:', result[1])
  ```

- **Làm Cho Chuỗi Thời Gian Ổn Định:** Nếu chuỗi không ổn định, thực hiện việc lấy sai phân.

  ```python
  df['Close_diff'] = df['Close'].diff().dropna()
  ```

**Bước 2: Điều Chỉnh Tham Số (p, d, q)**

- **p (AutoRegressive order):** Số lượng giá trị tự hồi quy.
- **d (Integrated order):** Số lần lấy sai phân để làm ổn định chuỗi.
- **q (Moving Average order):** Số lượng sai số trung bình động.

Sử dụng biểu đồ ACF (AutoCorrelation Function) và PACF (Partial AutoCorrelation Function) để xác định các giá trị p và q.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

plot_acf(df['Close_diff'].dropna(), lags=20)
plot_pacf(df['Close_diff'].dropna(), lags=20)
plt.show()
```

**Bước 3: Huấn Luyện Mô Hình ARIMA**

Sau khi xác định các tham số p, d, q, huấn luyện mô hình ARIMA.

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Close'], order=(p, d, q))
model_fit = model.fit()
print(model_fit.summary())
```

**Bước 4: Dự Báo và Đánh Giá Mô Hình**

Sử dụng mô hình đã huấn luyện để dự báo và đánh giá hiệu suất dựa trên các chỉ số như RMSE, MAE.

```python
forecast = model_fit.forecast(steps=30)
plt.figure(figsize=(10,6))
plt.plot(df['Close'], label='Actual')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.show()
```

#### 5.2.2. Áp Dụng Mô Hình Ridge Regression

**Bước 1: Chuẩn Bị Dữ Liệu**

Trước khi huấn luyện mô hình, dữ liệu cần được chuẩn hóa để đảm bảo rằng tất cả các đặc trưng đều nằm trong cùng một khoảng giá trị.

```python
from sklearn.preprocessing import MinMaxScaler

scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

### 5.3. Mô hình LSTM

#### 5.3.1. Giới Thiệu về LSTM

**LSTM (Long Short-Term Memory)** là một loại mạng nơ-ron hồi tiếp (Recurrent Neural Network - RNN) được thiết kế để giải quyết vấn đề **vanishing gradient**, cho phép mô hình ghi nhớ thông tin trong khoảng thời gian dài. LSTM thích hợp cho các bài toán dự báo chuỗi thời gian phức tạp như dự đoán giá cổ phiếu.

#### 5.3.2. Giải Thích Đơn Giản về LSTM

**Định Nghĩa Đơn Giản:**

- **LSTM** là một thuật toán thông minh trong lĩnh vực Trí tuệ Nhân tạo (AI), được sử dụng để **hiểu và dự đoán các chuỗi sự kiện theo thời gian**.
- Nó giống như một cuốn sổ tay thần kỳ, có thể **ghi nhớ những điều quan trọng trong quá khứ** và **quên những điều không quan trọng**, để đưa ra dự đoán chính xác hơn.

**Ví Dụ Đơn Giản:**

1. **Ký Ức của Bạn về Một Bài Kiểm Tra:**
   - Bạn đang học để thi. Đầu tiên, bạn nhớ rất nhiều thứ, nhưng sau một thời gian, bạn quên đi các chi tiết không quan trọng (ví dụ: màu bút bạn dùng), nhưng vẫn giữ lại những gì cần thiết (công thức toán).
   - LSTM hoạt động tương tự: nó quyết định nên nhớ hay quên thông tin cũ, dựa vào độ quan trọng của chúng.

2. **Dự Đoán Giá Cổ Phiếu:**
   - Để dự đoán giá cổ phiếu ngày mai, ta không chỉ dựa vào giá hôm nay mà còn cả **xu hướng dài hạn** (ví dụ, giá đang tăng đều trong 1 tháng) và **sự kiện gần đây** (ví dụ, thông báo từ công ty hôm qua). LSTM sẽ ghi nhớ cả hai loại thông tin này.

#### 5.3.3. Kiến Trúc của Mô Hình LSTM

Mô hình LSTM trong nghiên cứu này được xây dựng với kiến trúc gồm hai lớp LSTM và một lớp Dense để đưa ra dự đoán.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def build_model(self):
    """
    Build the LSTM model.
    """
    self.model = Sequential([
        LSTM(50, activation='relu', return_sequences=True, input_shape=(self.look_back, self.data.shape[1] - 2)),
        LSTM(50, activation='relu'),
        Dense(1)
    ])
    self.model.compile(optimizer='adam', loss='mse')
```

**Giải Thích:**

- **Lớp LSTM Đầu Tiên:**
  - Có 50 nơ-ron, sử dụng hàm kích hoạt ReLU.
  - `return_sequences=True` cho phép lớp này trả về toàn bộ chuỗi đầu ra để được xử lý bởi lớp LSTM tiếp theo.
  
- **Lớp LSTM Thứ Hai:**
  - Có 50 nơ-ron, sử dụng hàm kích hoạt ReLU.
  - `return_sequences=False` (mặc định) chỉ trả về trạng thái cuối cùng để được kết nối với lớp Dense.
  
- **Lớp Dense:**
  - Chỉ có một nơ-ron, nhằm mục đích dự đoán giá đóng cửa ngày tiếp theo.

**Các Kỹ Thuật Chính Quy (Regularization):**

- **Early Stopping:** Sử dụng để ngăn ngừa overfitting bằng cách dừng huấn luyện khi mô hình không cải thiện trên tập kiểm tra trong một số epoch nhất định.

  ```python
  from tensorflow.keras.callbacks import EarlyStopping

  early_stop = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
  ```

#### 5.3.4. Áp Dụng Mô Hình LSTM

**Bước 1: Tải và Tiền Xử Lý Dữ Liệu**

- Chuẩn hóa dữ liệu bằng `MinMaxScaler`.
- Tạo các chuỗi dữ liệu (look-back window) để mô hình LSTM có thể học từ các chuỗi giá trị trước đó.

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

# Tạo chuỗi dữ liệu
look_back = 10
X_sequences, y_sequences = [], []
for i in range(look_back, len(y_scaled)):
    X_sequences.append(X_scaled[i - look_back:i, :])
    y_sequences.append(y_scaled[i, 0])

X_sequences, y_sequences = np.array(X_sequences), np.array(y_sequences)
```

**Bước 2: Chia Dữ Liệu Thành Tập Huấn Luyện và Tập Kiểm Tra**

```python
train_size = int(len(X_sequences) * 0.75)
X_train, X_test = X_sequences[:train_size], X_sequences[train_size:]
y_train, y_test = y_sequences[:train_size], y_sequences[train_size:]
```

**Bước 3: Xây Dựng và Huấn Luyện Mô Hình**

```python
model = Sequential([
    LSTM(50, activation='relu', return_sequences=True, input_shape=(look_back, X_train.shape[2])),
    LSTM(50, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

early_stop = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[early_stop])
```

**Bước 4: Dự Báo và Đánh Giá Mô Hình**

```python
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Inverse transform
y_train_original = scaler_y.inverse_transform(y_train.reshape(-1, 1))
y_pred_train_original = scaler_y.inverse_transform(y_pred_train)
y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1))
y_pred_test_original = scaler_y.inverse_transform(y_pred_test)
```

Đánh giá mô hình bằng các chỉ số như R², MSE, RMSE, MAE, và MAPE.

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

r2_test = r2_score(y_test_original, y_pred_test_original)
mse_test = mean_squared_error(y_test_original, y_pred_test_original)
rmse_test = np.sqrt(mse_test)
mae_test = mean_absolute_error(y_test_original, y_pred_test_original)
mape_test = mean_absolute_percentage_error(y_test_original, y_pred_test_original) * 100

print(f"R²: {r2_test:.4f}")
print(f"MSE: {mse_test:.4f}")
print(f"RMSE: {rmse_test:.4f}")
print(f"MAE: {mae_test:.4f}")
print(f"MAPE: {mape_test:.2f}%")
```

#### 5.3.5. Đánh Giá Hiệu Suất Mô Hình

Để đánh giá hiệu suất của mô hình LSTM, chúng tôi sử dụng các chỉ số sau:

- **MSE (Mean Squared Error):** Đo lường trung bình bình phương sai số giữa giá trị thực và giá trị dự đoán.
- **RMSE (Root Mean Squared Error):** Đo lường sai số trung bình theo đơn vị giá trị mục tiêu.
- **MAE (Mean Absolute Error):** Đo lường sai số trung bình tuyệt đối.
- **MAPE (Mean Absolute Percentage Error):** Đo lường sai số trung bình tuyệt đối theo tỷ lệ phần trăm.

  - Triển khai và so sánh ba mô hình (Ridge Regression, LSTM, ARIMA) trên dữ liệu từ 10 công ty CNTT.
- **Phát hiện chính**:
  - Ridge Regression mang lại hiệu suất ổn định nhất; LSTM phù hợp cho dữ liệu phức tạp nhưng cần cải tiến thêm; ARIMA chỉ hiệu quả với dữ liệu tĩnh và ít biến động.

- **Đánh giá thời gian thực**:
  - Thực hiện dự đoán trong thời gian thực để kiểm tra tính ứng dụng của mô hình trong điều kiện thị trường thực tế.
