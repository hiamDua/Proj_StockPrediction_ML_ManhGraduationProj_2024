```python
import pandas as pd
import os

folder_dir = "D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/2_code_notebooks_TrainingModel/compare_RealVsPredict/"

# Set display options for better formatting
pd.set_option('display.float_format', lambda x: '{:.2f}'.format(x))
pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.colheader_justify', 'center')

# Get list of CSV files
csv_files = [f for f in os.listdir(folder_dir) if f.endswith('.csv')]

# Read and display each file
for file in csv_files:
    file_path = os.path.join(folder_dir, file)
    df = pd.read_csv(file_path)
    
    print(f"\nStock Price Comparison for {file}:")
    print("=" * 50)
    print(df.head(15).to_string(
        index=True,
        float_format=lambda x: '{:,.2f}'.format(x),
        justify='center',
        col_space=15
    ))
    print("=" * 50)
    print("\n")
```
