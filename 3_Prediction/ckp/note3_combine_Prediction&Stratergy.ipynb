### **1. Pair Trading truyền thống (đã thực hiện ở chương 4):**
- **Cách hoạt động:**
  1. Tính toán **spread** giữa hai cổ phiếu (chênh lệch giá).
  2. Chuẩn hóa spread (tính **z-score**) để xác định khi nào spread **quá lớn hoặc quá nhỏ** (outliers).
  3. Mở vị thế khi spread vượt ngưỡng ±2 (hoặc các ngưỡng khác do bạn chọn).
  4. Đóng vị thế khi spread quay lại gần mức bình thường (±0, ±0.5,...).

### **2. Kết hợp dự đoán (forecasting) với Pair Trading:**
Pair Trading không yêu cầu dự đoán giá trong các bước cơ bản, nhưng việc kết hợp với dự đoán có thể tăng hiệu quả giao dịch. Có hai cách để kết hợp:

#### **A. Dự đoán spread trực tiếp**
- **Ý tưởng:**
  - Thay vì chỉ dựa vào spread hiện tại, sử dụng mô hình dự đoán để **dự báo spread tương lai**.
  - Nếu mô hình dự đoán rằng spread sẽ mở rộng (hoặc thu hẹp), bạn có thể vào lệnh sớm hơn.
  
- **Cách làm:**
  1. Xây dựng mô hình dự đoán spread (Ridge Regression, LSTM, hoặc bất kỳ mô hình nào).
     - Đầu vào: Các đặc trưng từ FPT và CMG (giá đóng cửa, volume, biến động,...).
     - Đầu ra: Spread tương lai (giá trị trong 1, 5, hoặc 10 ngày tới).
  2. Sử dụng dự đoán này để điều chỉnh ngưỡng mở vị thế:
     - Nếu dự đoán spread tăng: Mở Short Spread.
     - Nếu dự đoán spread giảm: Mở Long Spread.

#### **B. Dự đoán xu hướng giá từng cổ phiếu**
- **Ý tưởng:**
  - Sử dụng mô hình dự đoán giá để dự báo xu hướng giá của FPT và CMG.
  - Xác định hành động Pair Trading dựa trên dự đoán này.
  
- **Cách làm:**
  1. Xây dựng mô hình dự đoán giá (Ridge Regression, LSTM, ARIMA,...).
     - Dự đoán giá của FPT và CMG trong tương lai.
  2. Tính **spread dự đoán**:
     \[
     \text{Spread dự đoán} = \text{Giá dự đoán FPT} - \text{Giá dự đoán CMG}
     \]
  3. Mở vị thế Pair Trading nếu spread dự đoán vượt ngưỡng:
     - **Short Spread**: Nếu spread dự đoán > ngưỡng.
     - **Long Spread**: Nếu spread dự đoán < -ngưỡng.

### **3. Kết hợp dự đoán và Pair Trading trong code**

#### **A. Dự đoán spread trực tiếp**
- **Bổ sung bước dự đoán spread:**

```python
def forecast_spread(data, model):
    """
    Forecast spread using a predictive model.
    """
    # Lấy đặc trưng từ dữ liệu
    features = data[['close_FPT', 'close_CMG', 'spread']].copy()
    features['spread_lag_1'] = features['spread'].shift(1)
    features['spread_lag_2'] = features['spread'].shift(2)
    features.dropna(inplace=True)

    # Dự đoán spread tương lai
    X = features[['spread', 'spread_lag_1', 'spread_lag_2']].values
    forecasted_spread = model.predict(X)
    
    # Thêm spread dự đoán vào dữ liệu
    data.loc[features.index, 'forecasted_spread'] = forecasted_spread
    return data
```

   - **Reversal Trading (VGI-VTL):** Sử dụng Ridge Regression để dự báo biến động giá của từng cổ phiếu, từ đó xác định spread dự báo.

   - **Sử dụng mô hình dự báo Ridge Regression (Chiến lược kết hợp):**
     - **Pair Trading:** Sử dụng dự báo spread từ Ridge Regression để điều chỉnh ngưỡng mở vị thế. Mở Short Spread nếu spread dự báo tăng và mở Long Spread nếu spread dự báo giảm.
