### **Giải thích cách thực hiện Pair Trading siêu đơn giản**

---

#### **Pair Trading là gì?**
Hãy tưởng tượng bạn có hai người bạn thân là **FPT** và **CMG**. Bình thường, hai bạn luôn đi gần nhau. Nhưng đôi khi, một bạn **chạy quá xa** còn bạn kia lại **đi chậm hơn**. Bạn biết rằng sau một lúc, hai bạn sẽ lại gần nhau.

- **Ý tưởng Pair Trading:** 
  - Khi hai bạn **đi xa nhau quá**, bạn đặt cược rằng hai bạn sẽ lại gần nhau:
    - **Mua bạn đi chậm (CMG)** và **bán bạn chạy nhanh (FPT)**.
  - Khi hai bạn **lại gần**, bạn bán những gì đã mua và mua lại những gì đã bán để kiếm lời.

---



#### **Cách làm chi tiết**

1. **Bước 1: Chuẩn bị dữ liệu**
   - Dùng bảng giá cổ phiếu của FPT và CMG trong 100 ngày gần nhất.
   - Tính "khoảng cách" giữa hai chúng ta (gọi là **spread**).

2. **Bước 2: Theo dõi khoảng cách (spread)**
   - Spread là khoảng cách giá giữa FPT và CMG:
     \[
     \text{Spread} = \text{Giá FPT} - \text{Giá CMG}
     \]
   - Khi khoảng cách này **quá lớn** (ví dụ: > 2 lần mức bình thường), chúng ta nghĩ rằng khoảng cách sẽ thu hẹp lại.

3. **Bước 3: Đặt cược khi khoảng cách quá lớn**
   - Nếu khoảng cách **quá cao**: chúng ta **bán FPT** (chúng ta chạy nhanh) và **mua CMG** (chúng ta đi chậm).
   - Nếu khoảng cách **quá thấp**: chúng ta **mua FPT** (chúng ta đi chậm) và **bán CMG** (chúng ta chạy nhanh).

4. **Bước 4: Đóng vị thế khi khoảng cách trở lại bình thường**
   - Khi khoảng cách quay về gần mức bình thường, chúng ta **bán những gì đã mua** và **mua lại những gì đã bán** để kiếm lời.

5. **Bước 5: Tính lời/lỗ**
   - Sau 100 ngày, chúng ta tổng kết xem đã kiếm được bao nhiêu từ việc đặt cược này.

---

#### **Kết quả chúng ta cần làm**
1. Theo dõi khoảng cách giữa FPT và CMG mỗi ngày.
2. Đặt cược khi khoảng cách **quá cao hoặc quá thấp**.
3. Tính tổng tiền lãi hoặc lỗ sau khi giao dịch trong 100 ngày.

- **Tính lời/lỗ:**
  - Lợi nhuận từ FPT = (90,000 - 85,000) × 1,000 = **5,000,000 đồng.**
  - Lợi nhuận từ CMG = (45,000 - 42,000) × 1,000 = **3,000,000 đồng.**

```python
import pandas as pd

# Load dữ liệu từ các file được tải lên
fpt_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/FPT_stock_data.csv'
cmg_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/CMG_stock_data.csv'

# Đọc dữ liệu FPT và CMG
fpt_data = pd.read_csv(fpt_file_path)
cmg_data = pd.read_csv(cmg_file_path)

# Hiển thị dữ liệu mẫu
fpt_data.head(), cmg_data.head()
```

## 2. Process data
- Kết hợp dữ liệu FPT và CMG theo ngày.
- Tính toán spread và z-score.
- Áp dụng logic giao dịch Pair Trading trong 100 ngày gần nhất.

- spread: Chênh lệch giá giữa FPT và CMG.   
- mean_spread: Giá trị trung bình của spread trong 30 ngày gần nhất.
- std_spread: Độ lệch chuẩn của spread trong 30 ngày gần nhất.
- z_score: Độ lệch chuẩn hóa của spread, dùng để xác định ngưỡng giao dịch.

```python
# Lấy cột 'time' và 'close', đổi tên cột 'close' để tránh nhầm lẫn
processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

# Định dạng cột 'time' và sắp xếp theo ngày
processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
processed_fpt_data.sort_values('time', inplace=True)
processed_cmg_data.sort_values('time', inplace=True)

# Kết hợp dữ liệu theo cột 'time'
data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

# Tính spread và các thông số khác
data['spread'] = data['close_FPT'] - data['close_CMG']
data['mean_spread'] = data['spread'].rolling(window=30).mean()
data['std_spread'] = data['spread'].rolling(window=30).std()
data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

# Lấy 100 ngày gần nhất
data_recent = data.tail(100)

# Hiển thị dữ liệu mẫu
data_recent
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Display the full dataset
print("\nFull dataset:")
print(data_recent.to_string())

# Optional: Reset display options to default after printing
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
pd.reset_option('display.width')
```

1. **`spread`**: Chênh lệch giá giữa FPT và CMG:
   $
   \text{spread} = \text{close\_FPT} - \text{close\_CMG}
   $
2. **`mean_spread`**: Giá trị trung bình của spread trong 30 ngày gần nhất (rolling mean).
3. **`std_spread`**: Độ lệch chuẩn của spread trong 30 ngày gần nhất (rolling std).
4. **`z_score`**: Giá trị chuẩn hóa của spread:
   $
   z\_score = \frac{\text{spread} - \text{mean\_spread}}{\text{std\_spread}}
   $


```python
# Khởi tạo biến
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Vòng lặp giao dịch
for index, row in data_recent.iterrows():
    z = row['z_score']

    # Mở vị thế
    if position == 0:
        if z > 2:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}")
        elif z < -2:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}")

    # Đóng vị thế
    elif position == 1 and z >= 0:
        # Đóng Long Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (exit_price_fpt - entry_price_fpt) * 1000 - (exit_price_cmg - entry_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

    elif position == -1 and z <= 0:
        # Đóng Short Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (entry_price_fpt - exit_price_fpt) * 1000 + (entry_price_cmg - exit_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

# Tổng kết
total_profit = sum(profits)
num_trades = len(profits)

print("\nKết quả giao dịch Pair Trading:")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
```

### **Vì sao cần giả lập với dữ liệu lịch sử?**

- **Giá đóng cửa của ngày tiếp theo:** Để tính toán lợi nhuận khi vị thế được đóng.

Điều này mô phỏng lại **quy trình thực tế**:
1. chúng ta quyết định giao dịch vào **cuối ngày hiện tại** dựa trên các tín hiệu (z-score).
2. Giao dịch sẽ được thực hiện vào ngày tiếp theo khi thị trường mở cửa.

# Chiến lược update

Để xây dựng chiến lược "xịn" và cải thiện hiệu quả giao dịch Pair Trading, chúng ta có thể kết hợp các yếu tố sau:

---

### **Đề xuất chiến lược cải thiện**
#### **1. Điều chỉnh ngưỡng Z-Score**
- Thay vì sử dụng ngưỡng Z-score cố định (±2), chúng ta có thể:
  - **Mở vị thế:** Khi Z-score vượt ±2.5 (tín hiệu rõ ràng hơn).
  - **Đóng vị thế:** Khi Z-score quay về ±0.5 (tăng xác suất đóng vị thế kịp thời).

---

#### **2. Thêm quy tắc cắt lỗ và chốt lời**
- **Chốt lời:** Đóng vị thế nếu lợi nhuận đạt ngưỡng kỳ vọng (ví dụ: **5,000,000 VND**).
- **Cắt lỗ:** Đóng vị thế nếu lỗ vượt ngưỡng chịu đựng (ví dụ: **3,000,000 VND**).

---

#### **3. Phân tích xu hướng (Trend Filtering)**
- Chỉ mở vị thế khi cả FPT và CMG không có xu hướng mạnh.
- Sử dụng các chỉ báo kỹ thuật như:
  - **RSI (Relative Strength Index):** Đảm bảo không giao dịch khi thị trường quá mua hoặc quá bán.
  - **Moving Averages:** Đảm bảo spread đang dao động quanh mức trung bình.

---

#### **4. Tăng thời gian phân tích**
- Sử dụng **rolling window dài hơn** (60 hoặc 90 ngày) để tính `mean_spread` và `std_spread` nhằm giảm nhiễu.

---

#### **5. Mô phỏng chiến lược**
Chúng ta sẽ thực hiện lại chiến lược với các điều chỉnh trên để kiểm tra kết quả:

1. **Mở vị thế khi Z > ±2.5.**
2. **Đóng vị thế khi Z quay về ±0.5 hoặc đạt ngưỡng cắt lỗ/chốt lời.**
3. **Tính toán lại lợi nhuận.**

```python
# Khởi tạo biến cho chiến lược cải tiến
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Tham số chiến lược
z_open_threshold = 2.5  # Ngưỡng mở vị thế
z_close_threshold = 0.5  # Ngưỡng đóng vị thế
profit_target = 5000000  # Chốt lời (5 triệu VND)
loss_limit = -3000000    # Cắt lỗ (3 triệu VND)

# Vòng lặp giao dịch trong 100 ngày gần nhất
for index, row in recent_data.iterrows():
    z = row['z_score']

    # Mở vị thế nếu không có vị thế
    if position == 0:
        if z > z_open_threshold:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}")
        elif z < -z_open_threshold:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}")

    # Đóng vị thế nếu đã mở
    elif position == 1:  # Long Spread
        current_profit = (row['close_FPT'] - entry_price_fpt) * 1000 - (row['close_CMG'] - entry_price_cmg) * 1000
        if z >= z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
            # Đóng Long Spread
            exit_price_fpt = row['close_FPT']
            exit_price_cmg = row['close_CMG']
            profits.append(current_profit)
            print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}, Lợi nhuận: {current_profit:.2f} VND")
            position = 0

    elif position == -1:  # Short Spread
        current_profit = (entry_price_fpt - row['close_FPT']) * 1000 + (entry_price_cmg - row['close_CMG']) * 1000
        if z <= -z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
            # Đóng Short Spread
            exit_price_fpt = row['close_FPT']
            exit_price_cmg = row['close_CMG']
            profits.append(current_profit)
            print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}, Lợi nhuận: {current_profit:.2f} VND")
            position = 0

# Tổng kết giao dịch
total_profit = sum(profits)
num_trades = len(profits)

# Kết quả tổng hợp
print("\nKết quả giao dịch Pair Trading (Chiến lược cải tiến):")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
```

  - Phản ánh mức giá ổn định nhất sau cả một ngày giao dịch.

- **Ý tưởng Pair Trading:** 
  - Khi hai bạn **đi xa nhau quá**, bạn đặt cược rằng hai bạn sẽ lại gần nhau:
    - **Mua bạn đi chậm (CMG)** và **bán bạn chạy nhanh (FPT)**.

## 2. Process data
- Kết hợp dữ liệu FPT và CMG theo ngày.
- Tính toán spread và z-score.
- Áp dụng logic giao dịch Pair Trading trong 100 ngày gần nhất.

- mean_spread: Giá trị trung bình của spread trong 30 ngày gần nhất.

- z_score: Độ lệch chuẩn hóa của spread, dùng để xác định ngưỡng giao dịch.

```python
# Lấy cột 'time' và 'close', đổi tên cột 'close' để tránh nhầm lẫn
processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

# Định dạng cột 'time' và sắp xếp theo ngày
processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
processed_fpt_data.sort_values('time', inplace=True)
processed_cmg_data.sort_values('time', inplace=True)

# Kết hợp dữ liệu theo cột 'time'
data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

# Tính spread và các thông số khác
data['spread'] = data['close_FPT'] - data['close_CMG']
data['mean_spread'] = data['spread'].rolling(window=30).mean()
data['std_spread'] = data['spread'].rolling(window=30).std()
data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

# Lấy 100 ngày gần nhất
data_recent = data.tail(100)

# Hiển thị dữ liệu mẫu
data_recent
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Display the full dataset
print("\nFull dataset:")
print(data_recent.to_string())

# Optional: Reset display options to default after printing
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
pd.reset_option('display.width')
```

1. **`spread`**: Chênh lệch giá giữa FPT và CMG:
   $
   \text{spread} = \text{close\_FPT} - \text{close\_CMG}
   $
2. **`mean_spread`**: Giá trị trung bình của spread trong 30 ngày gần nhất (rolling mean).
3. **`std_spread`**: Độ lệch chuẩn của spread trong 30 ngày gần nhất (rolling std).
4. **`z_score`**: Giá trị chuẩn hóa của spread:
   $
   z\_score = \frac{\text{spread} - \text{mean\_spread}}{\text{std\_spread}}
   $


```python
# Khởi tạo biến
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Vòng lặp giao dịch
for index, row in data_recent.iterrows():
    z = row['z_score']

    # Mở vị thế
    if position == 0:
        if z > 2:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}")
        elif z < -2:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}")

    # Đóng vị thế
    elif position == 1 and z >= 0:
        # Đóng Long Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (exit_price_fpt - entry_price_fpt) * 1000 - (exit_price_cmg - entry_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

    elif position == -1 and z <= 0:
        # Đóng Short Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (entry_price_fpt - exit_price_fpt) * 1000 + (entry_price_cmg - exit_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

# Tổng kết
total_profit = sum(profits)
num_trades = len(profits)

print("\nKết quả giao dịch Pair Trading:")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
```

```python
# Dựa trên dữ liệu, tiến hành cập nhật để sử dụng giá "open" cho ngày hôm sau.

# Lấy cột cần thiết từ dữ liệu và đổi tên để tránh nhầm lẫn
processed_fpt_data = fpt_data[['time', 'close', 'open']].rename(
    columns={'close': 'close_FPT', 'open': 'open_FPT'}
)
processed_cmg_data = cmg_data[['time', 'close', 'open']].rename(
    columns={'close': 'close_CMG', 'open': 'open_CMG'}
)

# Định dạng cột 'time' và sắp xếp theo ngày
processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
processed_fpt_data.sort_values('time', inplace=True)
processed_cmg_data.sort_values('time', inplace=True)

# Kết hợp dữ liệu theo cột 'time'
data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

# Tính spread và các thông số khác
data['spread'] = data['close_FPT'] - data['close_CMG']
data['mean_spread'] = data['spread'].rolling(window=30).mean()
data['std_spread'] = data['spread'].rolling(window=30).std()
data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

# Shift cột 'open' lên một ngày để mô phỏng giá mở cửa ngày hôm sau
data['next_open_FPT'] = data['open_FPT'].shift(-1)
data['next_open_CMG'] = data['open_CMG'].shift(-1)

# Xóa các hàng cuối cùng vì không có giá mở cửa ngày hôm sau
data = data[:-1]

# Lấy 100 ngày gần nhất
data_recent = data.tail(100)

# Khởi tạo biến
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Vòng lặp giao dịch
for index, row in data_recent.iterrows():
    z = row['z_score']

    # Mở vị thế
    if position == 0:
        if z > 2:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}")
        elif z < -2:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}")

    # Đóng vị thế
    elif position == 1 and z >= 0:
        # Đóng Long Spread
        exit_price_fpt = row['next_open_FPT']
        exit_price_cmg = row['next_open_CMG']
        profit = (exit_price_fpt - entry_price_fpt) * 1000 - (exit_price_cmg - entry_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

    elif position == -1 and z <= 0:
        # Đóng Short Spread
        exit_price_fpt = row['next_open_FPT']
        exit_price_cmg = row['next_open_CMG']
        profit = (entry_price_fpt - exit_price_fpt) * 1000 + (entry_price_cmg - exit_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

# Tổng kết
total_profit = sum(profits)
num_trades = len(profits)

# Kết quả tổng hợp
print("\nKết quả giao dịch Pair Trading (Sử dụng giá mở cửa ngày hôm sau):")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
```

## Nếu mua cái này bán cái kia hôm nay thì mai bán và mua luôn à hay sao ?? Có đúng là Pair Trading không?

### **1. Giao dịch trong ngày (mua và bán trong cùng ngày):**
- Nếu bạn **mua và bán cổ phiếu trong cùng ngày giao dịch**, điều này được gọi là **Day Trading (Giao dịch trong ngày)**.
- **Đặc điểm:**
  - Không giữ vị thế qua đêm, giúp tránh được rủi ro từ biến động qua đêm (**overnight gap**).
  - Yêu cầu tập trung cao để theo dõi giá trong ngày.
  - Phổ biến trong các chiến lược ngắn hạn.

---

### **2. Giao dịch mua hôm nay và bán ngay hôm sau (không chờ z-score thoả mãn):**
- **Không được xem là Pair Trading đúng nghĩa nếu bạn chưa đủ tín hiệu để đóng vị thế (như z-score về 0).**
- **Tuy nhiên, điều này vẫn có thể chấp nhận được trong thực tế**, nhưng với các lưu ý sau:

#### **Lợi ích:**
1. **Phản ứng nhanh với tín hiệu mới:**
   - Nếu thị trường biến động mạnh qua đêm, việc đóng vị thế ngay hôm sau có thể giúp giảm rủi ro.
   
2. **Không phụ thuộc quá nhiều vào mô hình:**
   - Nếu spread không có xu hướng quay về mức trung bình, bạn có thể chốt lời/lỗ nhanh chóng mà không chờ tín hiệu.

#### **Rủi ro:**
1. **Mất tính nhất quán của chiến lược:**
   - Pair Trading dựa trên giả định rằng spread sẽ quay về mức trung bình. Bán ngay hôm sau khi chưa đủ tín hiệu có thể làm giảm hiệu quả của chiến lược.

2. **Dễ bỏ lỡ cơ hội:**
   - Nếu bạn bán quá sớm, spread có thể thu hẹp vào những ngày tiếp theo, dẫn đến lợi nhuận tiềm năng bị bỏ lỡ.

---

### **3. Điều kiện chấp nhận được:**
- Nếu bạn có quy tắc giao dịch rõ ràng, như:
  - Đóng vị thế sau một khoảng thời gian cố định (ví dụ: 1 hoặc 2 ngày) để giảm rủi ro.
  - Chấp nhận mức lợi nhuận/lỗ tối thiểu thay vì chờ tín hiệu hoàn hảo.
  
- **Trong thực tế:**
  - Hành động này có thể phù hợp với các nhà giao dịch muốn giảm rủi ro hoặc không muốn giữ vị thế qua đêm quá lâu.
