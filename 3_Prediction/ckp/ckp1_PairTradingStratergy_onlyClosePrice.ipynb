### **Giải thích cách thực hiện Pair Trading siêu đơn giản**

---

#### **Pair Trading là gì?**
Hãy tưởng tượng bạn có hai người bạn thân là **FPT** và **CMG**. Bình thường, hai bạn luôn đi gần nhau. Nhưng đôi khi, một bạn **chạy quá xa** còn bạn kia lại **đi chậm hơn**. Bạn biết rằng sau một lúc, hai bạn sẽ lại gần nhau.

- **Ý tưởng Pair Trading:** 
  - Khi hai bạn **đi xa nhau quá**, bạn đặt cược rằng hai bạn sẽ lại gần nhau:
    - **Mua bạn đi chậm (CMG)** và **bán bạn chạy nhanh (FPT)**.
  - Khi hai bạn **lại gần**, bạn bán những gì đã mua và mua lại những gì đã bán để kiếm lời.

---



#### **Cách làm chi tiết**

1. **Bước 1: Chuẩn bị dữ liệu**
   - Dùng bảng giá cổ phiếu của FPT và CMG trong 100 ngày gần nhất.
   - Tính "khoảng cách" giữa hai chúng ta (gọi là **spread**).

2. **Bước 2: Theo dõi khoảng cách (spread)**
   - Spread là khoảng cách giá giữa FPT và CMG:
     \[
     \text{Spread} = \text{Giá FPT} - \text{Giá CMG}
     \]
   - Khi khoảng cách này **quá lớn** (ví dụ: > 2 lần mức bình thường), chúng ta nghĩ rằng khoảng cách sẽ thu hẹp lại.

3. **Bước 3: Đặt cược khi khoảng cách quá lớn**
   - Nếu khoảng cách **quá cao**: chúng ta **bán FPT** (chúng ta chạy nhanh) và **mua CMG** (chúng ta đi chậm).
   - Nếu khoảng cách **quá thấp**: chúng ta **mua FPT** (chúng ta đi chậm) và **bán CMG** (chúng ta chạy nhanh).

4. **Bước 4: Đóng vị thế khi khoảng cách trở lại bình thường**
   - Khi khoảng cách quay về gần mức bình thường, chúng ta **bán những gì đã mua** và **mua lại những gì đã bán** để kiếm lời.

5. **Bước 5: Tính lời/lỗ**
   - Sau 100 ngày, chúng ta tổng kết xem đã kiếm được bao nhiêu từ việc đặt cược này.

---

#### **Kết quả chúng ta cần làm**
1. Theo dõi khoảng cách giữa FPT và CMG mỗi ngày.
2. Đặt cược khi khoảng cách **quá cao hoặc quá thấp**.
3. Tính tổng tiền lãi hoặc lỗ sau khi giao dịch trong 100 ngày.

---

#### **Ví dụ đơn giản**

- **Ngày 1:**
  - Giá FPT = 90,000 đồng, Giá CMG = 42,000 đồng.
  - Spread = 90,000 - 42,000 = 48,000 đồng.
  - Spread lớn hơn mức bình thường, chúng ta quyết định:
    - **Bán 1,000 cổ phiếu FPT.**
    - **Mua 1,000 cổ phiếu CMG.**

- **Ngày 10:**
  - Giá FPT = 85,000 đồng, Giá CMG = 45,000 đồng.
  - Spread = 85,000 - 45,000 = 40,000 đồng (trở về bình thường).
  - chúng ta đóng vị thế:
    - **Mua lại 1,000 cổ phiếu FPT với giá 85,000 đồng.**
    - **Bán 1,000 cổ phiếu CMG với giá 45,000 đồng.**

- **Tính lời/lỗ:**
  - Lợi nhuận từ FPT = (90,000 - 85,000) × 1,000 = **5,000,000 đồng.**
  - Lợi nhuận từ CMG = (45,000 - 42,000) × 1,000 = **3,000,000 đồng.**
  - **Tổng lợi nhuận: 8,000,000 đồng.**

## 1. Load data

````python
import pandas as pd

# Load dữ liệu từ các file được tải lên
fpt_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/FPT_stock_data.csv'
cmg_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/CMG_stock_data.csv'

# Đọc dữ liệu FPT và CMG
fpt_data = pd.read_csv(fpt_file_path)
cmg_data = pd.read_csv(cmg_file_path)

# Hiển thị dữ liệu mẫu
fpt_data.head(), cmg_data.head()
````

## 2. Process data
- Kết hợp dữ liệu FPT và CMG theo ngày.
- Tính toán spread và z-score.
- Áp dụng logic giao dịch Pair Trading trong 100 ngày gần nhất.

- spread: Chênh lệch giá giữa FPT và CMG.   
- mean_spread: Giá trị trung bình của spread trong 30 ngày gần nhất.
- std_spread: Độ lệch chuẩn của spread trong 30 ngày gần nhất.
- z_score: Độ lệch chuẩn hóa của spread, dùng để xác định ngưỡng giao dịch.

````python
# Lấy cột 'time' và 'close', đổi tên cột 'close' để tránh nhầm lẫn
processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

# Định dạng cột 'time' và sắp xếp theo ngày
processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
processed_fpt_data.sort_values('time', inplace=True)
processed_cmg_data.sort_values('time', inplace=True)

# Kết hợp dữ liệu theo cột 'time'
data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

# Tính spread và các thông số khác
data['spread'] = data['close_FPT'] - data['close_CMG']
data['mean_spread'] = data['spread'].rolling(window=30).mean()
data['std_spread'] = data['spread'].rolling(window=30).std()
data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

# Lấy 100 ngày gần nhất
recent_data = data.tail(100)

# Hiển thị dữ liệu mẫu
recent_data
````

````python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Display the full dataset
print("\nFull dataset:")
print(recent_data.to_string())

# Optional: Reset display options to default after printing
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
pd.reset_option('display.width')
````

1. **`spread`**: Chênh lệch giá giữa FPT và CMG:
   $
   \text{spread} = \text{close\_FPT} - \text{close\_CMG}
   $
2. **`mean_spread`**: Giá trị trung bình của spread trong 30 ngày gần nhất (rolling mean).
3. **`std_spread`**: Độ lệch chuẩn của spread trong 30 ngày gần nhất (rolling std).
4. **`z_score`**: Giá trị chuẩn hóa của spread:
   $
   z\_score = \frac{\text{spread} - \text{mean\_spread}}{\text{std\_spread}}
   $


````python
# Khởi tạo biến
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Vòng lặp giao dịch
for index, row in recent_data.iterrows():
    z = row['z_score']

    # Mở vị thế
    if position == 0:
        if z > 2:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}")
        elif z < -2:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}")

    # Đóng vị thế
    elif position == 1 and z >= 0:
        # Đóng Long Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (exit_price_fpt - entry_price_fpt) * 1000 - (exit_price_cmg - entry_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

    elif position == -1 and z <= 0:
        # Đóng Short Spread
        exit_price_fpt = row['close_FPT']
        exit_price_cmg = row['close_CMG']
        profit = (entry_price_fpt - exit_price_fpt) * 1000 + (entry_price_cmg - exit_price_cmg) * 1000
        profits.append(profit)
        print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}, Lợi nhuận: {profit:.2f} VND")
        position = 0

# Tổng kết
total_profit = sum(profits)
num_trades = len(profits)

print("\nKết quả giao dịch Pair Trading:")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
````

## Cách làm hiện tại đúng chưa?

Khi áp dụng chiến lược **Pair Trading** (hoặc bất kỳ chiến lược giao dịch nào), chúng ta chỉ có thể dựa trên **dữ liệu lịch sử và giá hiện tại** để đưa ra quyết định mua/bán. Giá của ngày hôm sau là **không biết trước**, và chúng ta chỉ có thể dự đoán dựa trên:

1. **Dữ liệu hiện tại:** Các thông số như **spread**, **z-score**, và xu hướng thị trường.
2. **Logic chiến lược:** Quy tắc giao dịch dựa vào tín hiệu rõ ràng.

---

### **Vì sao cần giả lập với dữ liệu lịch sử?**
Trong giả lập này, chúng ta sử dụng dữ liệu lịch sử với:
- **Giá đóng cửa của ngày hiện tại:** Là thông tin để quyết định mở hoặc đóng vị thế.
- **Giá đóng cửa của ngày tiếp theo:** Để tính toán lợi nhuận khi vị thế được đóng.

Điều này mô phỏng lại **quy trình thực tế**:
1. chúng ta quyết định giao dịch vào **cuối ngày hiện tại** dựa trên các tín hiệu (z-score).
2. Giao dịch sẽ được thực hiện vào ngày tiếp theo khi thị trường mở cửa.

---

### **Điều này ảnh hưởng như thế nào?**
- Trong thực tế, vì chúng ta không biết giá của ngày tiếp theo, các quyết định chỉ dựa trên dữ liệu **lịch sử và hiện tại**. Do đó:
  - **Lợi nhuận thực tế** có thể khác so với giả lập, đặc biệt khi có biến động lớn vào ngày hôm sau.
  - Cần sử dụng các công cụ dự báo hoặc phân tích kỹ thuật để giảm rủi ro.

Nếu chúng ta muốn mô phỏng "thực tế hơn", chúng ta có thể:
- Dựa vào giá **mở cửa** ngày hôm sau để thực hiện giao dịch thay vì giá đóng cửa.

## Chiến lược cải thiện

Để xây dựng chiến lược "xịn" và cải thiện hiệu quả giao dịch Pair Trading, chúng ta có thể kết hợp các yếu tố sau:

---

### **Đề xuất chiến lược cải thiện**
#### **1. Điều chỉnh ngưỡng Z-Score**
- Thay vì sử dụng ngưỡng Z-score cố định (±2), chúng ta có thể:
  - **Mở vị thế:** Khi Z-score vượt ±2.5 (tín hiệu rõ ràng hơn).
  - **Đóng vị thế:** Khi Z-score quay về ±0.5 (tăng xác suất đóng vị thế kịp thời).

---

#### **2. Thêm quy tắc cắt lỗ và chốt lời**
- **Chốt lời:** Đóng vị thế nếu lợi nhuận đạt ngưỡng kỳ vọng (ví dụ: **5,000,000 VND**).
- **Cắt lỗ:** Đóng vị thế nếu lỗ vượt ngưỡng chịu đựng (ví dụ: **3,000,000 VND**).

---

#### **3. Phân tích xu hướng (Trend Filtering)**
- Chỉ mở vị thế khi cả FPT và CMG không có xu hướng mạnh.
- Sử dụng các chỉ báo kỹ thuật như:
  - **RSI (Relative Strength Index):** Đảm bảo không giao dịch khi thị trường quá mua hoặc quá bán.
  - **Moving Averages:** Đảm bảo spread đang dao động quanh mức trung bình.

---

#### **4. Tăng thời gian phân tích**
- Sử dụng **rolling window dài hơn** (60 hoặc 90 ngày) để tính `mean_spread` và `std_spread` nhằm giảm nhiễu.

---

#### **5. Mô phỏng chiến lược**
Chúng ta sẽ thực hiện lại chiến lược với các điều chỉnh trên để kiểm tra kết quả:

1. **Mở vị thế khi Z > ±2.5.**
2. **Đóng vị thế khi Z quay về ±0.5 hoặc đạt ngưỡng cắt lỗ/chốt lời.**
3. **Tính toán lại lợi nhuận.**

---




````python
# Cải tiến thêm thông tin chi tiết khi mở hoặc đóng vị thế
position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
entry_price_fpt = 0
entry_price_cmg = 0
profits = []

# Tham số chiến lược
z_open_threshold = 2.5  # Ngưỡng mở vị thế
z_close_threshold = 0.5  # Ngưỡng đóng vị thế
profit_target = 5000000  # Chốt lời (5 triệu VND)
loss_limit = -3000000    # Cắt lỗ (3 triệu VND)

# Vòng lặp giao dịch trong 100 ngày gần nhất
for index, row in recent_data.iterrows():
    z = row['z_score']

    # Mở vị thế nếu không có vị thế
    if position == 0:
        if z > z_open_threshold:
            # Mở Short Spread: Bán FPT, Mua CMG
            position = -1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Short Spread: Bán FPT {entry_price_fpt}, Mua CMG {entry_price_cmg}. Lý do: z_score ({z:.2f}) vượt ngưỡng {z_open_threshold}.")
        elif z < -z_open_threshold:
            # Mở Long Spread: Mua FPT, Bán CMG
            position = 1
            entry_price_fpt = row['close_FPT']
            entry_price_cmg = row['close_CMG']
            print(f"{row['time']} - Mở Long Spread: Mua FPT {entry_price_fpt}, Bán CMG {entry_price_cmg}. Lý do: z_score ({z:.2f}) dưới ngưỡng {-z_open_threshold}.")

    # Đóng vị thế nếu đã mở
    elif position == 1:  # Long Spread
        current_profit = (row['close_FPT'] - entry_price_fpt) * 1000 - (row['close_CMG'] - entry_price_cmg) * 1000
        if z >= z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
            # Đóng Long Spread
            exit_price_fpt = row['close_FPT']
            exit_price_cmg = row['close_CMG']
            reason = ""
            if z >= z_close_threshold:
                reason = f"z_score ({z:.2f}) đạt ngưỡng đóng vị thế ({z_close_threshold})."
            elif current_profit >= profit_target:
                reason = f"Đạt mức chốt lời {profit_target} VND."
            elif current_profit <= loss_limit:
                reason = f"Chạm mức cắt lỗ {loss_limit} VND."
            profits.append(current_profit)
            print(f"{row['time']} - Đóng Long Spread: Bán FPT {exit_price_fpt}, Mua CMG {exit_price_cmg}. Lợi nhuận: {current_profit:.2f} VND. Lý do: {reason}")
            position = 0

    elif position == -1:  # Short Spread
        current_profit = (entry_price_fpt - row['close_FPT']) * 1000 + (entry_price_cmg - row['close_CMG']) * 1000
        if z <= -z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
            # Đóng Short Spread
            exit_price_fpt = row['close_FPT']
            exit_price_cmg = row['close_CMG']
            reason = ""
            if z <= -z_close_threshold:
                reason = f"z_score ({z:.2f}) đạt ngưỡng đóng vị thế ({-z_close_threshold})."
            elif current_profit >= profit_target:
                reason = f"Đạt mức chốt lời {profit_target} VND."
            elif current_profit <= loss_limit:
                reason = f"Chạm mức cắt lỗ {loss_limit} VND."
            profits.append(current_profit)
            print(f"{row['time']} - Đóng Short Spread: Mua FPT {exit_price_fpt}, Bán CMG {exit_price_cmg}. Lợi nhuận: {current_profit:.2f} VND. Lý do: {reason}")
            position = 0

# Tổng kết giao dịch
total_profit = sum(profits)
num_trades = len(profits)

# Kết quả tổng hợp
print("\nKết quả giao dịch Pair Trading (Chiến lược cải tiến):")
print(f"Số lần giao dịch: {num_trades}")
print(f"Tổng lợi nhuận: {total_profit:.2f} VND")
````

````python
import pandas as pd


class PairTradingStrategy:
    def __init__(self, fpt_file_path, cmg_file_path, recent_days, strategy_params):
        """
        Initialize the trading strategy with data files, days, and strategy parameters.
        """
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.recent_days = recent_days
        self.strategy_params = strategy_params
        self.data = None
        self.profits = []

    def load_data(self):
        """
        Load and preprocess the data.
        """
        # Read data
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        # Process relevant columns
        processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        # Convert time to datetime and sort
        processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
        processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
        processed_fpt_data.sort_values('time', inplace=True)
        processed_cmg_data.sort_values('time', inplace=True)

        # Merge datasets
        data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

        # Calculate spread and z-score
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=30).mean()
        data['std_spread'] = data['spread'].rolling(window=30).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        # Filter recent days
        self.data = data.tail(self.recent_days)

    def execute_strategy(self, strategy_name):
        """
        Execute a trading strategy.
        """
        position = 0  # 0: No position, 1: Long Spread, -1: Short Spread
        entry_price_fpt = 0
        entry_price_cmg = 0
        self.profits = []  # Reset profits for each strategy

        print(f"\nExecuting {strategy_name} on last {self.recent_days} days...\n")
        params = self.strategy_params[strategy_name]

        for index, row in self.data.iterrows():
            z = row['z_score']

            # Open position
            if position == 0:
                if z > params['z_open_threshold']:
                    position = -1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    print(f"{row['time']} - Open Short Spread: Sell FPT {entry_price_fpt}, Buy CMG {entry_price_cmg}. Reason: z_score ({z:.2f}) > {params['z_open_threshold']}.")
                elif z < -params['z_open_threshold']:
                    position = 1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    print(f"{row['time']} - Open Long Spread: Buy FPT {entry_price_fpt}, Sell CMG {entry_price_cmg}. Reason: z_score ({z:.2f}) < {-params['z_open_threshold']}.")

            # Close position
            elif position == 1:  # Long Spread
                current_profit = (row['close_FPT'] - entry_price_fpt) * 1000 - (row['close_CMG'] - entry_price_cmg) * 1000
                if z >= params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    reason = self.get_close_reason(z, current_profit, params)
                    print(f"{row['time']} - Close Long Spread: Sell FPT {row['close_FPT']}, Buy CMG {row['close_CMG']}. Profit: {current_profit:.2f} VND. Reason: {reason}")
                    self.profits.append(current_profit)
                    position = 0

            elif position == -1:  # Short Spread
                current_profit = (entry_price_fpt - row['close_FPT']) * 1000 + (entry_price_cmg - row['close_CMG']) * 1000
                if z <= -params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    reason = self.get_close_reason(z, current_profit, params)
                    print(f"{row['time']} - Close Short Spread: Buy FPT {row['close_FPT']}, Sell CMG {row['close_CMG']}. Profit: {current_profit:.2f} VND. Reason: {reason}")
                    self.profits.append(current_profit)
                    position = 0

        total_profit = sum(self.profits)
        print(f"\nResults for {strategy_name}:")
        print(f"Number of Trades: {len(self.profits)}")
        print(f"Total Profit: {total_profit:.2f} VND\n")

    @staticmethod
    def get_close_reason(z, profit, params):
        """
        Determine the reason for closing the position.
        """
        if z >= params['z_close_threshold']:
            return f"z_score ({z:.2f}) >= {params['z_close_threshold']}."
        elif z <= -params['z_close_threshold']:
            return f"z_score ({z:.2f}) <= {-params['z_close_threshold']}."
        elif profit >= params['profit_target']:
            return f"Reached profit target ({params['profit_target']} VND)."
        elif profit <= params['loss_limit']:
            return f"Hit loss limit ({params['loss_limit']} VND)."
        return "Other condition."

# File paths
fpt_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/FPT_stock_data.csv'
cmg_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/CMG_stock_data.csv'

# Strategy parameters
strategy_params = {
    "strategy_1": {
        "z_open_threshold": 2.0,  # Ngưỡng mở vị thế
        "z_close_threshold": 0.0,  # Ngưỡng đóng vị thế
        "profit_target": float('inf'),  # Không chốt lời
        "loss_limit": float('-inf'),  # Không cắt lỗ
    },
    "strategy_2": {
        "z_open_threshold": 2.5,
        "z_close_threshold": 0.5,
        "profit_target": 5000000,
        "loss_limit": -3000000,
    }
}


# Initialize strategy tester
recent_days = 600  # Change to 100 or 1000 as needed
tester = PairTradingStrategy(fpt_file_path, cmg_file_path, recent_days, strategy_params)

# Run tests
tester.load_data()
tester.execute_strategy("strategy_1")
tester.execute_strategy("strategy_2")
````

   - Điều này giúp xác định xem chiến lược có hoạt động ổn định hay không, ngay cả khi dữ liệu đầu vào thay đổi.

2. **Đánh giá tính hiệu quả:**

   - Nếu chiến lược chỉ hiệu quả ở một số tập dữ liệu nhất định, cần xem xét điều chỉnh hoặc loại bỏ các tham số không phù hợp.

3. **Lựa chọn khoảng thời gian phù hợp:**
   - Giả sử chiến lược hoạt động tốt nhất với `recent_days = 800`, đây sẽ là độ dài dữ liệu tối ưu để thực hiện giao dịch thực tế.

### **Cách sử dụng `recent_days_range`**

- **Ví dụ:**
  ```python
  recent_days_range = range(100, 1001, 100)
  ```

  - ...

  1. Tải dữ liệu tương ứng.
  2. Áp dụng bộ tham số hiện tại.
  3. Đo lường hiệu quả (lợi nhuận, số giao dịch, tỷ lệ thắng, v.v.).

### **Kết luận**

  - **Khoảng cuối:** 2023-09-01 đến 2024-01-01.

## Ý tưởng 2: Chia chunk_size thì nên lấy 70% train, 30% test => Tìm ra bộ parameters tốt nhất

Việc chia dữ liệu thành 70% train và 30% test để tìm bộ tham số tốt nhất là một cách hợp lý và thường được áp dụng trong các bài toán tối ưu chiến lược giao dịch. Dưới đây là cách triển khai:

---

### **Ý tưởng thực hiện**
1. **Chia dữ liệu**:
   - Sử dụng 70% dữ liệu để "huấn luyện" (train), tìm kiếm bộ tham số tốt nhất.
   - Sử dụng 30% dữ liệu để "kiểm tra" (test), đánh giá hiệu quả của bộ tham số tìm được.

2. **Tìm tham số tốt nhất**:
   - Dựa trên tập dữ liệu train, thử nghiệm các bộ tham số (`z_open_threshold`, `z_close_threshold`, `profit_target`, `loss_limit`) và tìm bộ tham số tối ưu.

3. **Kiểm tra trên tập test**:
   - Áp dụng bộ tham số tối ưu đã tìm được trên tập test.
   - So sánh hiệu quả thực tế (lợi nhuận, số lần giao dịch).


### **Ý nghĩa**
1. **Train phase (70%)**:
   - Dùng để tìm bộ tham số tối ưu, dựa trên dữ liệu trong quá khứ.

2. **Test phase (30%)**:
   - Kiểm tra bộ tham số trên dữ liệu mới, để đánh giá tính ổn định và hiệu quả.

### **Ưu điểm**
- Đảm bảo rằng chiến lược không bị "quá khớp" (overfitting) vào dữ liệu train.
- Đánh giá tính thực tiễn của chiến lược khi áp dụng vào dữ liệu mới.

````python
class HyperparameterTuner:
    def __init__(self, fpt_file_path, cmg_file_path, train_ratio=0.7):
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.train_ratio = train_ratio
        self.train_data = None
        self.test_data = None

    def load_data(self):
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
        processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
        processed_fpt_data.sort_values('time', inplace=True)
        processed_cmg_data.sort_values('time', inplace=True)

        data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=30).mean()
        data['std_spread'] = data['spread'].rolling(window=30).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        split_idx = int(len(data) * self.train_ratio)
        self.train_data = data.iloc[:split_idx]
        self.test_data = data.iloc[split_idx:]

    def execute_strategy(self, data, params):
        position = 0
        entry_price_fpt = 0
        entry_price_cmg = 0
        profits = []

        for _, row in data.iterrows():
            z = row['z_score']

            if position == 0:
                if z > params['z_open_threshold']:
                    position = -1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                elif z < -params['z_open_threshold']:
                    position = 1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']

            elif position == 1:
                current_profit = (row['close_FPT'] - entry_price_fpt) * 1000 - (row['close_CMG'] - entry_price_cmg) * 1000
                if z >= params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    profits.append(current_profit)
                    position = 0

            elif position == -1:
                current_profit = (entry_price_fpt - row['close_FPT']) * 1000 + (entry_price_cmg - row['close_CMG']) * 1000
                if z <= -params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    profits.append(current_profit)
                    position = 0

        return profits

    def tune_parameters(self, z_open_range, z_close_range, profit_targets, loss_limits):
        self.load_data()
        results = []

        for z_open, z_close, profit_target, loss_limit in product(z_open_range, z_close_range, profit_targets, loss_limits):
            params = {
                "z_open_threshold": z_open,
                "z_close_threshold": z_close,
                "profit_target": profit_target,
                "loss_limit": loss_limit,
            }
            profits = self.execute_strategy(self.train_data, params)
            total_profit = sum(profits)
            num_trades = len(profits)
            avg_profit_per_trade = total_profit / num_trades if num_trades > 0 else 0

            results.append({
                "z_open_threshold": z_open,
                "z_close_threshold": z_close,
                "profit_target": profit_target,
                "loss_limit": loss_limit,
                "total_profit": total_profit,
                "number_of_trades": num_trades,
                "average_profit_per_trade": avg_profit_per_trade,
            })

        return pd.DataFrame(results)

    def evaluate_on_test(self, best_params):
        profits = self.execute_strategy(self.test_data, best_params)
        total_profit = sum(profits)
        num_trades = len(profits)
        avg_profit_per_trade = total_profit / num_trades if num_trades > 0 else 0

        return {
            "total_profit": total_profit,
            "number_of_trades": num_trades,
            "average_profit_per_trade": avg_profit_per_trade,
        }


# Hyperparameter ranges
z_open_range = [2.0, 2.5, 3.0]
z_close_range = [0.0, 0.5, 1.0]
profit_targets = [5000000, 7000000, 10000000]
loss_limits = [-3000000, -5000000, -7000000]

# Initialize tuner and tune parameters
tuner = HyperparameterTuner(fpt_file_path, cmg_file_path)
results_df = tuner.tune_parameters(z_open_range, z_close_range, profit_targets, loss_limits)

# Select best parameters
results_df.sort_values(by="total_profit", ascending=False, inplace=True)
best_params = results_df.iloc[0].to_dict()

# Evaluate on test data
test_results = tuner.evaluate_on_test(best_params)

# Display results
print("\nBest Parameters (Train):", best_params)
print("\nEvaluation on Test Data:", test_results)
````

# Triển khai Hyperparameter Tuning Chuẩn Ý TƯỞNG TIME SERIES

### **Triển khai Hyperparameter Tuning Chuẩn Ý TƯỞNG TIME SERIES**

Trong bối cảnh dữ liệu **Time Series**, cách tiếp cận chuẩn cần đảm bảo rằng:
1. **Không vi phạm tính thời gian**: Tức là khi tối ưu hóa hoặc kiểm định, dữ liệu tương lai không được phép "rò rỉ" vào dữ liệu quá khứ.
2. **Tối ưu hóa trên dữ liệu quá khứ và kiểm tra trên dữ liệu tương lai**: Đây là nguyên tắc quan trọng để đảm bảo rằng chiến lược có thể áp dụng được trong thực tế.
3. **Tối ưu hóa tham số phù hợp với bản chất tuần tự của dữ liệu**: Đảm bảo các bộ tham số được kiểm tra tính ổn định qua các giai đoạn thị trường khác nhau.

---


### **Bảng Tóm Tắt 3 Chiến Lược Triển Khai Hyperparameter Tuning Chuẩn Ý TƯỞNG TIME SERIES**

| **Chiến lược**             | **Nguyên tắc chính**                                                                 | **Ưu điểm**                                                                                     | **Nhược điểm**                                                                                 | **Khi nào dùng**                                                                                                                   |
|----------------------------|-------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------|
| **Walk-Forward Validation** | - Chia dữ liệu thành các tập **train** và **test** theo thời gian.                   | - Không vi phạm tính thời gian.<br>- Kiểm tra tính ổn định trên nhiều giai đoạn.<br>- Rõ ràng. | - Yêu cầu tính toán cao do tối ưu nhiều lần.<br>- Phức tạp khi triển khai.                   | - Khi cần đánh giá chiến lược một cách khách quan nhất.<br>- Đảm bảo không "rò rỉ" dữ liệu tương lai vào quá khứ.                 |
| **Expanding Window**        | - Tập **train** mở rộng dần qua thời gian.<br>- Test trên dữ liệu mới nhất.         | - Tận dụng được nhiều dữ liệu nhất khi tối ưu.<br>- Thích hợp cho thị trường thay đổi chậm.     | - Tập train giai đoạn đầu nhỏ, khó tối ưu.<br>- Cần đảm bảo dữ liệu mới hơn phản ánh xu hướng. | - Khi bạn tin rằng xu hướng dài hạn của dữ liệu có tác động lớn hơn so với biến động ngắn hạn.                                     |
| **Rolling Window**          | - Cả **train** và **test** có kích thước cố định.<br>- Cửa sổ trượt qua thời gian.  | - Phù hợp nếu dữ liệu cũ ít quan trọng.<br>- Đảm bảo tính liên tục giữa train và test.         | - Có thể bỏ qua xu hướng dài hạn.<br>- Cần chọn đúng kích thước cửa sổ.                      | - Khi muốn kiểm tra tính hiệu quả của chiến lược trong các giai đoạn ngắn và tránh dùng dữ liệu quá cũ.<br>- Thị trường biến động. |

---

### **Cách Triển Khai Mỗi Chiến Lược**
| **Chiến lược**             | **Train Data**               | **Test Data**                | **Cách di chuyển cửa sổ**                                                                                          |
|----------------------------|-----------------------------|------------------------------|--------------------------------------------------------------------------------------------------------------------|
| **Walk-Forward Validation** | - Tăng dần từ 1 → \(N\).     | - Ngay sau \(N+1 → N+M\).     | - Dịch chuyển toàn bộ train-test đến giai đoạn tiếp theo.                                                          |
| **Expanding Window**        | - Mở rộng từ 1 → \(N\).      | - Ngay sau \(N+1 → N+M\).     | - Train tăng kích thước qua mỗi bước, test giữ cố định.                                                            |
| **Rolling Window**          | - Cố định \(N-M → N\).       | - Ngay sau \(N+1 → N+M\).     | - Train và test có kích thước cố định, dịch chuyển đồng thời qua thời gian.                                        |

---

### **Khi Nào Dùng?**

| **Chiến lược**             | **Khi nào dùng?**                                                                                                           |
|----------------------------|----------------------------------------------------------------------------------------------------------------------------|
| **Walk-Forward Validation** | - Đánh giá khách quan và toàn diện.<br>- Kiểm tra chiến lược dài hạn.<br>- Dữ liệu thị trường thay đổi đa dạng qua các giai đoạn. |
| **Expanding Window**        | - Tận dụng dữ liệu tối đa.<br>- Dữ liệu thay đổi dần dần theo thời gian.<br>- Xu hướng dài hạn ảnh hưởng mạnh đến kết quả.     |
| **Rolling Window**          | - Thị trường có nhiều biến động ngắn hạn.<br>- Dữ liệu cũ ít giá trị.<br>- Cần đánh giá hiệu quả chiến lược trong từng giai đoạn ngắn. |

---

### **Hướng Dẫn Lựa Chọn**
- **Nếu bạn cần đảm bảo tính tổng quát và chính xác cao:** Dùng **Walk-Forward Validation**.
- **Nếu bạn muốn tận dụng toàn bộ dữ liệu và xu hướng dài hạn:** Dùng **Expanding Window**.
- **Nếu bạn cần đánh giá các chiến lược trong giai đoạn ngắn và muốn giảm tác động của dữ liệu cũ:** Dùng **Rolling Window**.

````python
import pandas as pd
import numpy as np
from itertools import product
import matplotlib.pyplot as plt

class PairTradingStrategy:
    def __init__(self, fpt_file_path, cmg_file_path, window=30):
        """
        Initialize the Pair Trading Strategy with file paths and rolling window size.
        """
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.window = window
        self.data = None

    def load_and_process_data(self):
        """
        Load and preprocess the stock data.
        """
        # Đọc dữ liệu từ file CSV
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        # Chọn các cột cần thiết và đổi tên để tránh trùng lặp
        processed_fpt = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        # Chuyển đổi cột 'time' thành kiểu datetime và sắp xếp
        processed_fpt['time'] = pd.to_datetime(processed_fpt['time'])
        processed_cmg['time'] = pd.to_datetime(processed_cmg['time'])
        processed_fpt.sort_values('time', inplace=True)
        processed_cmg.sort_values('time', inplace=True)

        # Kết hợp dữ liệu theo ngày
        data = pd.merge(processed_fpt, processed_cmg, on='time')

        # Tính spread, mean_spread, std_spread và z_score
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=self.window).mean()
        data['std_spread'] = data['spread'].rolling(window=self.window).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        # Loại bỏ các dòng có giá trị NaN do rolling window
        data.dropna(inplace=True)

        self.data = data.reset_index(drop=True)
        return self.data

    def plot_spread(self):
        """
        Plot the spread and z-score over time.
        """
        plt.figure(figsize=(14, 7))

        plt.subplot(2, 1, 1)
        plt.plot(self.data['time'], self.data['spread'], label='Spread')
        plt.plot(self.data['time'], self.data['mean_spread'], label='Mean Spread')
        plt.fill_between(self.data['time'], 
                         self.data['mean_spread'] + 2*self.data['std_spread'], 
                         self.data['mean_spread'] - 2*self.data['std_spread'], 
                         color='grey', alpha=0.3, label='±2 Std Dev')
        plt.title('Spread giữa FPT và CMG')
        plt.legend()

        plt.subplot(2, 1, 2)
        plt.plot(self.data['time'], self.data['z_score'], label='Z-Score')
        plt.axhline(2, color='red', linestyle='--', label='Z > 2')
        plt.axhline(-2, color='green', linestyle='--', label='Z < -2')
        plt.title('Z-Score của Spread')
        plt.legend()

        plt.tight_layout()
        plt.show()

    def execute_trading_strategy(self, z_open_threshold, z_close_threshold, profit_target=np.inf, loss_limit=-np.inf, calc_profit_with_open=False):
        """
        Execute the Pair Trading strategy with given parameters.
        
        Parameters:
            z_open_threshold (float): Ngưỡng mở vị thế.
            z_close_threshold (float): Ngưỡng đóng vị thế.
            profit_target (float): Mức chốt lời.
            loss_limit (float): Mức cắt lỗ.
            calc_profit_with_open (bool): Nếu True, tính lợi nhuận bằng Open của ngày đóng vị thế.
        Returns:
            profits (list): Danh sách lợi nhuận từ các giao dịch.
            trade_details (list): Chi tiết từng giao dịch.
            total_profit (float): Tổng lợi nhuận.
            num_trades (int): Số lần giao dịch.
        """
        position = 0  # 0: Không có vị thế, 1: Long Spread, -1: Short Spread
        entry_price_fpt = 0
        entry_price_cmg = 0
        profits = []
        trade_details = []

        for i in range(len(self.data)):
            row = self.data.iloc[i]
            z = row['z_score']
            date = row['time']
            close_fpt = row['close_FPT']
            close_cmg = row['close_CMG']
            # Lấy giá Open của ngày tiếp theo nếu có
            if (i+1) < len(self.data):
                next_row = self.data.iloc[i+1]
                open_fpt = next_row['close_FPT']
                open_cmg = next_row['close_CMG']
            else:
                open_fpt = close_fpt
                open_cmg = close_cmg

            # Mở vị thế
            if position == 0:
                if z > z_open_threshold:
                    position = -1  # Short Spread
                    entry_price_fpt = close_fpt
                    entry_price_cmg = close_cmg
                    trade_details.append({
                        'open_date': date,
                        'strategy': 'Short Spread',
                        'open_price_fpt': close_fpt,
                        'open_price_cmg': close_cmg,
                        'reason': f'z_score ({z:.2f}) > {z_open_threshold}'
                    })
                elif z < -z_open_threshold:
                    position = 1  # Long Spread
                    entry_price_fpt = close_fpt
                    entry_price_cmg = close_cmg
                    trade_details.append({
                        'open_date': date,
                        'strategy': 'Long Spread',
                        'open_price_fpt': close_fpt,
                        'open_price_cmg': close_cmg,
                        'reason': f'z_score ({z:.2f}) < {-z_open_threshold}'
                    })

            # Đóng vị thế Long Spread
            elif position == 1:
                current_profit = (close_fpt - entry_price_fpt) * 1000 - (close_cmg - entry_price_cmg) * 1000
                if z >= z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
                    if calc_profit_with_open and (i+1) < len(self.data):
                        exit_price_fpt = open_fpt
                        exit_price_cmg = open_cmg
                        current_profit = (exit_price_fpt - entry_price_fpt) * 1000 - (exit_price_cmg - entry_price_cmg) * 1000
                    else:
                        exit_price_fpt = close_fpt
                        exit_price_cmg = close_cmg

                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': self.data.iloc[i+1]['time'] if (i+1) < len(self.data) else date,
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score >= z_close_threshold' if z >= z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0

            # Đóng vị thế Short Spread
            elif position == -1:
                current_profit = (entry_price_fpt - close_fpt) * 1000 + (entry_price_cmg - close_cmg) * 1000
                if z <= -z_close_threshold or current_profit >= profit_target or current_profit <= loss_limit:
                    if calc_profit_with_open and (i+1) < len(self.data):
                        exit_price_fpt = open_fpt
                        exit_price_cmg = open_cmg
                        current_profit = (entry_price_fpt - exit_price_fpt) * 1000 + (entry_price_cmg - exit_price_cmg) * 1000
                    else:
                        exit_price_fpt = close_fpt
                        exit_price_cmg = close_cmg

                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': self.data.iloc[i+1]['time'] if (i+1) < len(self.data) else date,
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score <= -z_close_threshold' if z <= -z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0

        # Tổng hợp kết quả
        total_profit = sum(profits)
        num_trades = len(profits)
        return profits, trade_details, total_profit, num_trades

class HyperparameterTuner:
    def __init__(self, strategy, param_grid, window=30, step=100):
        """
        Initialize the Hyperparameter Tuner.
        
        Parameters:
            strategy (PairTradingStrategy): Instance of PairTradingStrategy.
            param_grid (dict): Dictionary chứa các giá trị tham số cần thử nghiệm.
            window (int): Kích thước rolling window để tính spread.
            step (int): Số ngày để dịch chuyển cửa sổ.
        """
        self.strategy = strategy
        self.param_grid = param_grid
        self.window = window
        self.step = step

    def walk_forward_validation(self):
        """
        Perform Walk-Forward Validation to tune hyperparameters.
        
        Returns:
            best_params (dict): Bộ tham số tốt nhất dựa trên tổng lợi nhuận.
            all_results (list): Danh sách tất cả các kết quả từ các bộ tham số.
        """
        # Tổng hợp tất cả các bộ tham số
        keys, values = zip(*self.param_grid.items())
        param_combinations = [dict(zip(keys, v)) for v in product(*values)]

        all_results = []

        # Xác định số lượng bước
        total_days = len(self.strategy.data)
        for params in param_combinations:
            z_open = params['z_open_threshold']
            z_close = params['z_close_threshold']
            profit_target = params['profit_target']
            loss_limit = params['loss_limit']

            total_profit = 0
            num_trades = 0

            # Thực hiện từng bước Walk-Forward
            for start in range(0, total_days - self.step, self.step):
                end = start + self.step
                train_data = self.strategy.data.iloc[start:end]
                tester = PairTradingStrategy('', '', window=self.window)
                tester.data = train_data.copy()

                profits, _, tp, nt = tester.execute_trading_strategy(
                    z_open, z_close, profit_target, loss_limit, calc_profit_with_open=False
                )
                total_profit += tp
                num_trades += nt

            # Lưu kết quả
            all_results.append({
                'z_open_threshold': z_open,
                'z_close_threshold': z_close,
                'profit_target': profit_target,
                'loss_limit': loss_limit,
                'total_profit': total_profit,
                'number_of_trades': num_trades,
                'average_profit_per_trade': total_profit / num_trades if num_trades > 0 else 0
            })

        # Chuyển kết quả thành DataFrame
        results_df = pd.DataFrame(all_results)

        # Tìm bộ tham số có tổng lợi nhuận cao nhất
        best_params = results_df.loc[results_df['total_profit'].idxmax()].to_dict()

        return best_params, results_df
````

````python
def test_strategy_with_recent_days(strategy, best_params, recent_days_list):
    """
    Test the trading strategy with different recent days.

    Parameters:
        strategy (PairTradingStrategy): Chiến lược giao dịch.
        best_params (dict): Bộ tham số tốt nhất đã tối ưu hóa.
        recent_days_list (list): Danh sách số ngày cần kiểm tra.

    Returns:
        results (list): Kết quả giao dịch cho từng số ngày.
    """
    results = []

    for recent_days in recent_days_list:
        # Lấy dữ liệu n ngày gần nhất
        recent_data = strategy.data.tail(recent_days)
        strategy.data = recent_data  # Cập nhật dữ liệu vào chiến lược

        # Áp dụng chiến lược giao dịch với bộ tham số tốt nhất
        profits, trade_details, total_profit, num_trades = strategy.execute_trading_strategy(
            best_params['z_open_threshold'],
            best_params['z_close_threshold'],
            best_params['profit_target'],
            best_params['loss_limit'],
            calc_profit_with_open=False  # Hoặc True nếu muốn tính lợi nhuận dựa trên giá Open
        )

        # Lưu kết quả
        results.append({
            'recent_days': recent_days,
            'num_trades': num_trades,
            'total_profit': total_profit,
            'average_profit_per_trade': total_profit / num_trades if num_trades > 0 else 0,
            'details': trade_details  # Lưu chi tiết giao dịch
        })

    return results


# Danh sách số ngày cần kiểm tra
recent_days_list = [100, 200, 300, 400, 500]

# Gọi hàm kiểm tra
results = test_strategy_with_recent_days(strategy, best_params, recent_days_list)

# Hiển thị kết quả
for result in results:
    print(f"\n--- Kết quả cho {result['recent_days']} ngày gần nhất ---")
    print(f"Số lần giao dịch: {result['num_trades']}")
    print(f"Tổng lợi nhuận: {result['total_profit']:.2f} VND")
    print(f"Lợi nhuận trung bình mỗi giao dịch: {result['average_profit_per_trade']:.2f} VND")
    print("Chi tiết giao dịch:")
    for trade in result['details']:
        print(trade)
````
