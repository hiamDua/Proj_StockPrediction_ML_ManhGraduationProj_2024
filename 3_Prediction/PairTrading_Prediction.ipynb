### **2. Kết hợp dự đoán (forecasting) với Pair Trading:**
Pair Trading không yêu cầu dự đoán giá trong các bước cơ bản, nhưng việc kết hợp với dự đoán có thể tăng hiệu quả giao dịch. Có hai cách để kết hợp:

#### **A. Dự đoán spread trực tiếp**
- **Ý tưởng:**
  - Thay vì chỉ dựa vào spread hiện tại, sử dụng mô hình dự đoán để **dự báo spread tương lai**.
  - Nếu mô hình dự đoán rằng spread sẽ mở rộng (hoặc thu hẹp), bạn có thể vào lệnh sớm hơn.
  
- **Cách làm:**
  1. Xây dựng mô hình dự đoán spread (Ridge Regression, LSTM, hoặc bất kỳ mô hình nào).
     - Đầu vào: Các đặc trưng từ FPT và CMG (giá đóng cửa, volume, biến động,...).
     - Đầu ra: Spread tương lai (giá trị trong 1, 5, hoặc 10 ngày tới).
  2. Sử dụng dự đoán này để điều chỉnh ngưỡng mở vị thế:
     - Nếu dự đoán spread tăng: Mở Short Spread.
     - Nếu dự đoán spread giảm: Mở Long Spread.

#### **B. Dự đoán xu hướng giá từng cổ phiếu**
- **Ý tưởng:**
  - Sử dụng mô hình dự đoán giá để dự báo xu hướng giá của FPT và CMG.
  - Xác định hành động Pair Trading dựa trên dự đoán này.
  
- **Cách làm:**
  1. Xây dựng mô hình dự đoán giá (Ridge Regression, LSTM, ARIMA,...).
     - Dự đoán giá của FPT và CMG trong tương lai.
  2. Tính **spread dự đoán**:
     \[
     \text{Spread dự đoán} = \text{Giá dự đoán FPT} - \text{Giá dự đoán CMG}
     \]
  3. Mở vị thế Pair Trading nếu spread dự đoán vượt ngưỡng:
     - **Short Spread**: Nếu spread dự đoán > ngưỡng.
     - **Long Spread**: Nếu spread dự đoán < -ngưỡng.

```python
import pandas as pd


class PairTradingStrategy:
    def __init__(self, fpt_file_path, cmg_file_path, recent_days, strategy_params):
        """
        Initialize the trading strategy with data files, days, and strategy parameters.
        """
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.recent_days = recent_days
        self.strategy_params = strategy_params
        self.data = None
        self.profits = []

    def load_data(self):
        """
        Load and preprocess the data.
        """
        # Read data
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        # Process relevant columns
        processed_fpt_data = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg_data = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        # Convert time to datetime and sort
        processed_fpt_data['time'] = pd.to_datetime(processed_fpt_data['time'])
        processed_cmg_data['time'] = pd.to_datetime(processed_cmg_data['time'])
        processed_fpt_data.sort_values('time', inplace=True)
        processed_cmg_data.sort_values('time', inplace=True)

        # Merge datasets
        data = pd.merge(processed_fpt_data, processed_cmg_data, on='time')

        # Calculate spread and z-score
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=30).mean()
        data['std_spread'] = data['spread'].rolling(window=30).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        # Filter recent days
        self.data = data.tail(self.recent_days)

    def execute_strategy(self, strategy_name):
        """
        Execute a trading strategy.
        """
        position = 0  # 0: No position, 1: Long Spread, -1: Short Spread
        entry_price_fpt = 0
        entry_price_cmg = 0
        self.profits = []  # Reset profits for each strategy

        print(f"\nExecuting {strategy_name} on last {self.recent_days} days...\n")
        params = self.strategy_params[strategy_name]

        for index, row in self.data.iterrows():
            z = row['z_score']

            # Open position
            if position == 0:
                if z > params['z_open_threshold']:
                    position = -1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    print(f"{row['time']} - Open Short Spread: Sell FPT {entry_price_fpt}, Buy CMG {entry_price_cmg}. Reason: z_score ({z:.2f}) > {params['z_open_threshold']}.")
                elif z < -params['z_open_threshold']:
                    position = 1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    print(f"{row['time']} - Open Long Spread: Buy FPT {entry_price_fpt}, Sell CMG {entry_price_cmg}. Reason: z_score ({z:.2f}) < {-params['z_open_threshold']}.")

            # Close position
            elif position == 1:  # Long Spread
                current_profit = (row['close_FPT'] - entry_price_fpt) * 1000 - (row['close_CMG'] - entry_price_cmg) * 1000
                if z >= params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    reason = self.get_close_reason(z, current_profit, params)
                    print(f"{row['time']} - Close Long Spread: Sell FPT {row['close_FPT']}, Buy CMG {row['close_CMG']}. Profit: {current_profit:.2f} VND. Reason: {reason}")
                    self.profits.append(current_profit)
                    position = 0

            elif position == -1:  # Short Spread
                current_profit = (entry_price_fpt - row['close_FPT']) * 1000 + (entry_price_cmg - row['close_CMG']) * 1000
                if z <= -params['z_close_threshold'] or current_profit >= params['profit_target'] or current_profit <= params['loss_limit']:
                    reason = self.get_close_reason(z, current_profit, params)
                    print(f"{row['time']} - Close Short Spread: Buy FPT {row['close_FPT']}, Sell CMG {row['close_CMG']}. Profit: {current_profit:.2f} VND. Reason: {reason}")
                    self.profits.append(current_profit)
                    position = 0

        total_profit = sum(self.profits)
        print(f"\nResults for {strategy_name}:")
        print(f"Number of Trades: {len(self.profits)}")
        print(f"Total Profit: {total_profit:.2f} VND\n")

    @staticmethod
    def get_close_reason(z, profit, params):
        """
        Determine the reason for closing the position.
        """
        if z >= params['z_close_threshold']:
            return f"z_score ({z:.2f}) >= {params['z_close_threshold']}."
        elif z <= -params['z_close_threshold']:
            return f"z_score ({z:.2f}) <= {-params['z_close_threshold']}."
        elif profit >= params['profit_target']:
            return f"Reached profit target ({params['profit_target']} VND)."
        elif profit <= params['loss_limit']:
            return f"Hit loss limit ({params['loss_limit']} VND)."
        return "Other condition."

# File paths
fpt_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/FPT_stock_data.csv'
cmg_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/CMG_stock_data.csv'

# Strategy parameters
strategy_params = {
    "strategy_1": {
        "z_open_threshold": 2.0,  # Ngưỡng mở vị thế
        "z_close_threshold": 0.0,  # Ngưỡng đóng vị thế
        "profit_target": float('inf'),  # Không chốt lời
        "loss_limit": float('-inf'),  # Không cắt lỗ
    },
    "strategy_2": {
        "z_open_threshold": 2.5,
        "z_close_threshold": 0.5,
        "profit_target": 5000000,
        "loss_limit": -3000000,
    }
}


# Initialize strategy tester
recent_days = 600  # Change to 100 or 1000 as needed
tester = PairTradingStrategy(fpt_file_path, cmg_file_path, recent_days, strategy_params)

# Run tests
tester.load_data()
tester.execute_strategy("strategy_1")
tester.execute_strategy("strategy_2")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

class PairTradingStrategy:
    def __init__(self, fpt_file_path, cmg_file_path, window=30):
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.window = window
        self.data = None
        self.model = Ridge()
    
    def load_and_process_data(self):
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        processed_fpt = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        processed_fpt['time'] = pd.to_datetime(processed_fpt['time'])
        processed_cmg['time'] = pd.to_datetime(processed_cmg['time'])
        processed_fpt.sort_values('time', inplace=True)
        processed_cmg.sort_values('time', inplace=True)

        data = pd.merge(processed_fpt, processed_cmg, on='time')
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=self.window).mean()
        data['std_spread'] = data['spread'].rolling(window=self.window).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        data.dropna(inplace=True)
        self.data = data.reset_index(drop=True)
        return self.data
    
    def train_ridge_model(self):
        self.data['spread_lag1'] = self.data['spread'].shift(1)
        self.data['spread_lag2'] = self.data['spread'].shift(2)
        self.data['spread_lag3'] = self.data['spread'].shift(3)
        self.data['spread_lag4'] = self.data['spread'].shift(4)
        self.data['spread_lag5'] = self.data['spread'].shift(5)
        self.data.dropna(inplace=True)

        X = self.data[['spread_lag1', 'spread_lag2', 'spread_lag3', 'spread_lag4', 'spread_lag5']]
        y = self.data['spread']

        split_ratio = 0.7
        split_index = int(len(X) * split_ratio)
        X_train, X_test = X[:split_index], X[split_index:]
        y_train, y_test = y[:split_index], y[split_index:]

        self.model.fit(X_train, y_train)
        y_train_pred = self.model.predict(X_train)
        y_test_pred = self.model.predict(X_test)

        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
        test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

        print(f"Train RMSE: {train_rmse:.2f}")
        print(f"Test RMSE: {test_rmse:.2f}")

        self.data['spread_pred'] = self.model.predict(X)
        self.data.dropna(inplace=True)
    
    def execute_trading_strategy(self, z_open_threshold, z_close_threshold, profit_target=np.inf, loss_limit=-np.inf):
        position = 0
        entry_price_fpt = 0
        entry_price_cmg = 0
        profits = []
        trade_details = []

        for i in range(len(self.data)-1):
            row = self.data.iloc[i]
            next_row = self.data.iloc[i+1]
            z = row['z_score']
            spread_pred = row['spread_pred']

            if position == 0:
                if spread_pred > 0:
                    position = -1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    trade_details.append({
                        'open_date': row['time'],
                        'strategy': 'Short Spread',
                        'open_price_fpt': entry_price_fpt,
                        'open_price_cmg': entry_price_cmg,
                        'reason': f'Spread_pred ({spread_pred:.2f}) > 0'
                    })
                elif spread_pred < 0:
                    position = 1
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    trade_details.append({
                        'open_date': row['time'],
                        'strategy': 'Long Spread',
                        'open_price_fpt': entry_price_fpt,
                        'open_price_cmg': entry_price_cmg,
                        'reason': f'Spread_pred ({spread_pred:.2f}) < 0'
                    })
            
            elif position == 1:
                current_profit = (next_row['close_FPT'] - entry_price_fpt) * 1000 - (next_row['close_CMG'] - entry_price_cmg) * 1000
                if (z >= z_close_threshold) or (current_profit >= profit_target) or (current_profit <= loss_limit):
                    if (i+1) < len(self.data):
                        exit_price_fpt = next_row['close_FPT']
                        exit_price_cmg = next_row['close_CMG']
                    else:
                        exit_price_fpt = row['close_FPT']
                        exit_price_cmg = row['close_CMG']
                    
                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': next_row['time'] if (i+1) < len(self.data) else row['time'],
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score >= z_close_threshold' if z >= z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0
            
            elif position == -1:
                current_profit = (entry_price_fpt - next_row['close_FPT']) * 1000 + (entry_price_cmg - next_row['close_CMG']) * 1000
                if (z <= -z_close_threshold) or (current_profit >= profit_target) or (current_profit <= loss_limit):
                    if (i+1) < len(self.data):
                        exit_price_fpt = next_row['close_FPT']
                        exit_price_cmg = next_row['close_CMG']
                    else:
                        exit_price_fpt = row['close_FPT']
                        exit_price_cmg = row['close_CMG']
                    
                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': next_row['time'] if (i+1) < len(self.data) else row['time'],
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score <= -z_close_threshold' if z <= -z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0

        total_profit = sum(profits)
        num_trades = len(profits)
        return profits, trade_details, total_profit, num_trades
    
    def report_trades(self, trade_details, total_profit, num_trades):
        print("\n### **Kết quả giao dịch Pair Trading với Ridge Regression**\n")
        print(f"- **Số lần giao dịch:** {num_trades} lần.")
        print(f"- **Tổng lợi nhuận:** {total_profit:.2f} VND.\n")
        print("#### **Chi tiết giao dịch:**\n")
        for idx, trade in enumerate(trade_details, 1):
            print(f"{idx}. **Giao dịch:** {trade['strategy']}")
            print(f"   - **Ngày mở vị thế:** {trade['open_date'].date()}")
            print(f"   - **Giá mở FPT:** {trade['open_price_fpt']} VND.")
            print(f"   - **Giá mở CMG:** {trade['open_price_cmg']} VND.")
            print(f"   - **Ngày đóng vị thế:** {trade['close_date'].date()}")
            print(f"   - **Giá đóng FPT:** {trade['close_price_fpt']} VND.")
            print(f"   - **Giá đóng CMG:** {trade['close_price_cmg']} VND.")
            print(f"   - **Lợi nhuận:** {trade['profit']:.2f} VND.")
            print(f"   - **Lý do đóng vị thế:** {trade['reason']}\n")

# --- Triển khai ---

# Đường dẫn tới file dữ liệu
fpt_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/FPT_stock_data.csv'
cmg_file_path = 'D:/OneDrive - Hanoi University of Science and Technology/GIT/MiniProj_StockPrediction_ML_SpManhGraduationProj_2024/data/raw20192024/CMG_stock_data.csv'

# Khởi tạo chiến lược và xử lý dữ liệu
strategy = PairTradingStrategy(fpt_file_path, cmg_file_path, window=30)
data = strategy.load_and_process_data()

# Huấn luyện mô hình Ridge Regression
strategy.train_ridge_model()

# Vẽ biểu đồ để kiểm tra
strategy.plot_spread()

# Áp dụng chiến lược giao dịch kết hợp Ridge Regression
# Các tham số chiến lược
z_open_threshold = 2.5  # Ngưỡng mở vị thế
z_close_threshold = 0.5  # Ngưỡng đóng vị thế
profit_target = 5000000  # Chốt lời 5 triệu VND
loss_limit = -3000000    # Cắt lỗ 3 triệu VND

profits, trade_details, total_profit, num_trades = strategy.execute_trading_strategy(
    z_open_threshold, 
    z_close_threshold, 
    profit_target, 
    loss_limit
)

# Báo cáo kết quả giao dịch
strategy.report_trades(trade_details, total_profit, num_trades)
```

#### **B. Dự đoán xu hướng giá từng cổ phiếu**
- **Ý tưởng:**
  - Sử dụng mô hình dự đoán giá để dự báo xu hướng giá của FPT và CMG.
  - Xác định hành động Pair Trading dựa trên dự đoán này.
  
- **Cách làm:**
  1. Xây dựng mô hình dự đoán giá (Ridge Regression, LSTM, ARIMA,...).
     - Dự đoán giá của FPT và CMG trong tương lai.
  2. Tính **spread dự đoán**:
     \[
     \text{Spread dự đoán} = \text{Giá dự đoán FPT} - \text{Giá dự đoán CMG}
     \]
  3. Mở vị thế Pair Trading nếu spread dự đoán vượt ngưỡng:
     - **Short Spread**: Nếu spread dự đoán > ngưỡng.
     - **Long Spread**: Nếu spread dự đoán < -ngưỡng.
|

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

class PairTradingStrategy:
    def __init__(self, fpt_file_path, cmg_file_path, window=30):
        self.fpt_file_path = fpt_file_path
        self.cmg_file_path = cmg_file_path
        self.window = window
        self.data = None
        self.model_fpt = Ridge()
        self.model_cmg = Ridge()
    
    def load_and_process_data(self):
        """
        Load and preprocess the stock data.
        """
        fpt_data = pd.read_csv(self.fpt_file_path)
        cmg_data = pd.read_csv(self.cmg_file_path)

        # Chọn các cột cần thiết và đổi tên để tránh trùng lặp
        processed_fpt = fpt_data[['time', 'close']].rename(columns={'close': 'close_FPT'})
        processed_cmg = cmg_data[['time', 'close']].rename(columns={'close': 'close_CMG'})

        # Chuyển đổi cột 'time' thành kiểu datetime và sắp xếp
        processed_fpt['time'] = pd.to_datetime(processed_fpt['time'])
        processed_cmg['time'] = pd.to_datetime(processed_cmg['time'])
        processed_fpt.sort_values('time', inplace=True)
        processed_cmg.sort_values('time', inplace=True)

        # Kết hợp dữ liệu theo ngày
        data = pd.merge(processed_fpt, processed_cmg, on='time')

        # Tính spread và các thông số khác
        data['spread'] = data['close_FPT'] - data['close_CMG']
        data['mean_spread'] = data['spread'].rolling(window=self.window).mean()
        data['std_spread'] = data['spread'].rolling(window=self.window).std()
        data['z_score'] = (data['spread'] - data['mean_spread']) / data['std_spread']

        # Loại bỏ các dòng có giá trị NaN do rolling window
        data.dropna(inplace=True)
        self.data = data.reset_index(drop=True)
        return self.data
    
    def train_ridge_models(self):
        """
        Train Ridge Regression models to predict future prices of FPT and CMG.
        """
        # Tạo các đặc trưng lagged cho FPT và CMG
        for lag in range(1, 6):
            self.data[f'close_FPT_lag{lag}'] = self.data['close_FPT'].shift(lag)
            self.data[f'close_CMG_lag{lag}'] = self.data['close_CMG'].shift(lag)
        
        self.data.dropna(inplace=True)

        # Định nghĩa X và y cho FPT
        features_fpt = [f'close_FPT_lag{lag}' for lag in range(1,6)]
        X_fpt = self.data[features_fpt]
        y_fpt = self.data['close_FPT']

        # Định nghĩa X và y cho CMG
        features_cmg = [f'close_CMG_lag{lag}' for lag in range(1,6)]
        X_cmg = self.data[features_cmg]
        y_cmg = self.data['close_CMG']

        # Chia dữ liệu thành train và test theo thời gian
        split_ratio = 0.7
        split_index = int(len(X_fpt) * split_ratio)

        X_train_fpt, X_test_fpt = X_fpt[:split_index], X_fpt[split_index:]
        y_train_fpt, y_test_fpt = y_fpt[:split_index], y_fpt[split_index:]

        X_train_cmg, X_test_cmg = X_cmg[:split_index], X_cmg[split_index:]
        y_train_cmg, y_test_cmg = y_cmg[:split_index], y_cmg[split_index:]

        # Huấn luyện mô hình Ridge Regression cho FPT
        self.model_fpt.fit(X_train_fpt, y_train_fpt)
        y_train_pred_fpt = self.model_fpt.predict(X_train_fpt)
        y_test_pred_fpt = self.model_fpt.predict(X_test_fpt)

        # Huấn luyện mô hình Ridge Regression cho CMG
        self.model_cmg.fit(X_train_cmg, y_train_cmg)
        y_train_pred_cmg = self.model_cmg.predict(X_train_cmg)
        y_test_pred_cmg = self.model_cmg.predict(X_test_cmg)

        # Tính toán độ chính xác
        train_rmse_fpt = np.sqrt(mean_squared_error(y_train_fpt, y_train_pred_fpt))
        test_rmse_fpt = np.sqrt(mean_squared_error(y_test_fpt, y_test_pred_fpt))
        print(f"FPT - Train RMSE: {train_rmse_fpt:.2f}")
        print(f"FPT - Test RMSE: {test_rmse_fpt:.2f}")

        train_rmse_cmg = np.sqrt(mean_squared_error(y_train_cmg, y_train_pred_cmg))
        test_rmse_cmg = np.sqrt(mean_squared_error(y_test_cmg, y_test_pred_cmg))
        print(f"CMG - Train RMSE: {train_rmse_cmg:.2f}")
        print(f"CMG - Test RMSE: {test_rmse_cmg:.2f}")

        # Dự báo giá trong toàn bộ dữ liệu
        self.data['close_FPT_pred'] = self.model_fpt.predict(X_fpt)
        self.data['close_CMG_pred'] = self.model_cmg.predict(X_cmg)
        self.data['spread_pred'] = self.data['close_FPT_pred'] - self.data['close_CMG_pred']
    
    def plot_spread(self):
        """
        Plot the spread and z-score over time.
        """
        plt.figure(figsize=(14, 10))

        # Plot Spread thực tế và Spread dự đoán
        plt.subplot(3, 1, 1)
        plt.plot(self.data['time'], self.data['spread'], label='Spread Thực Tế')
        plt.plot(self.data['time'], self.data['spread_pred'], label='Spread Dự Đoán', alpha=0.7)
        plt.plot(self.data['time'], self.data['mean_spread'], label='Mean Spread')
        plt.fill_between(self.data['time'], 
                         self.data['mean_spread'] + 2*self.data['std_spread'], 
                         self.data['mean_spread'] - 2*self.data['std_spread'], 
                         color='grey', alpha=0.3, label='±2 Std Dev')
        plt.title('Spread Thực Tế và Spread Dự Đoán giữa FPT và CMG')
        plt.legend()

        # Plot Z-Score thực tế
        plt.subplot(3, 1, 2)
        plt.plot(self.data['time'], self.data['z_score'], label='Z-Score Thực Tế')
        plt.axhline(2, color='red', linestyle='--', label='Z > 2')
        plt.axhline(-2, color='green', linestyle='--', label='Z < -2')
        plt.title('Z-Score Thực Tế của Spread')
        plt.legend()

        # Plot Spread Dự Đoán
        plt.subplot(3, 1, 3)
        plt.plot(self.data['time'], self.data['spread_pred'], label='Spread Dự Đoán', color='orange')
        plt.axhline(z_open_threshold, color='red', linestyle='--', label=f'Spread_pred > {z_open_threshold}')
        plt.axhline(-z_open_threshold, color='green', linestyle='--', label=f'Spread_pred < {-z_open_threshold}')
        plt.title('Spread Dự Đoán giữa FPT và CMG')
        plt.legend()

        plt.tight_layout()
        plt.show()
    
    def execute_trading_strategy(self, z_open_threshold, z_close_threshold, profit_target=np.inf, loss_limit=-np.inf):
        """
        Execute the Pair Trading strategy based on predicted spread.
        """
        position = 0  # 0: No position, 1: Long Spread, -1: Short Spread
        entry_price_fpt = 0
        entry_price_cmg = 0
        profits = []
        trade_details = []

        for i in range(len(self.data)-1):
            row = self.data.iloc[i]
            next_row = self.data.iloc[i+1]
            spread_pred = row['spread_pred']
            z = row['z_score']

            if position == 0:
                # Mở vị thế dựa trên spread dự đoán
                if spread_pred > z_open_threshold:
                    position = -1  # Short Spread
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    trade_details.append({
                        'open_date': row['time'],
                        'strategy': 'Short Spread',
                        'open_price_fpt': entry_price_fpt,
                        'open_price_cmg': entry_price_cmg,
                        'reason': f'Spread_pred ({spread_pred:.2f}) > {z_open_threshold}'
                    })
                elif spread_pred < -z_open_threshold:
                    position = 1  # Long Spread
                    entry_price_fpt = row['close_FPT']
                    entry_price_cmg = row['close_CMG']
                    trade_details.append({
                        'open_date': row['time'],
                        'strategy': 'Long Spread',
                        'open_price_fpt': entry_price_fpt,
                        'open_price_cmg': entry_price_cmg,
                        'reason': f'Spread_pred ({spread_pred:.2f}) < {-z_open_threshold}'
                    })

            elif position == 1:
                # Long Spread: Mua FPT, Bán CMG
                current_profit = (next_row['close_FPT'] - entry_price_fpt) * 1000 - (next_row['close_CMG'] - entry_price_cmg) * 1000
                if (z >= z_close_threshold) or (current_profit >= profit_target) or (current_profit <= loss_limit):
                    if (i+1) < len(self.data):
                        exit_price_fpt = next_row['close_FPT']
                        exit_price_cmg = next_row['close_CMG']
                    else:
                        exit_price_fpt = row['close_FPT']
                        exit_price_cmg = row['close_CMG']
                    
                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': next_row['time'] if (i+1) < len(self.data) else row['time'],
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score >= z_close_threshold' if z >= z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0

            elif position == -1:
                # Short Spread: Bán FPT, Mua CMG
                current_profit = (entry_price_fpt - next_row['close_FPT']) * 1000 + (entry_price_cmg - next_row['close_CMG']) * 1000
                if (z <= -z_close_threshold) or (current_profit >= profit_target) or (current_profit <= loss_limit):
                    if (i+1) < len(self.data):
                        exit_price_fpt = next_row['close_FPT']
                        exit_price_cmg = next_row['close_CMG']
                    else:
                        exit_price_fpt = row['close_FPT']
                        exit_price_cmg = row['close_CMG']
                    
                    profits.append(current_profit)
                    trade_details[-1].update({
                        'close_date': next_row['time'] if (i+1) < len(self.data) else row['time'],
                        'close_price_fpt': exit_price_fpt,
                        'close_price_cmg': exit_price_cmg,
                        'profit': current_profit,
                        'reason': 'z_score <= -z_close_threshold' if z <= -z_close_threshold else 
                                  'profit_target reached' if current_profit >= profit_target else 
                                  'loss_limit reached'
                    })
                    position = 0

        # Đóng vị thế còn mở nếu có
        if position != 0:
            last_row = self.data.iloc[-1]
            if position == 1:
                # Long Spread: Mua FPT, Bán CMG
                current_profit = (last_row['close_FPT'] - entry_price_fpt) * 1000 - (last_row['close_CMG'] - entry_price_cmg) * 1000
                profits.append(current_profit)
                trade_details[-1].update({
                    'close_date': last_row['time'],
                    'close_price_fpt': last_row['close_FPT'],
                    'close_price_cmg': last_row['close_CMG'],
                    'profit': current_profit,
                    'reason': 'End of data - đóng vị thế còn mở'
                })
            elif position == -1:
                # Short Spread: Bán FPT, Mua CMG
                current_profit = (entry_price_fpt - last_row['close_FPT']) * 1000 + (entry_price_cmg - last_row['close_CMG']) * 1000
                profits.append(current_profit)
                trade_details[-1].update({
                    'close_date': last_row['time'],
                    'close_price_fpt': last_row['close_FPT'],
                    'close_price_cmg': last_row['close_CMG'],
                    'profit': current_profit,
                    'reason': 'End of data - đóng vị thế còn mở'
                })

        total_profit = sum(profits)
        num_trades = len(profits)
        return profits, trade_details, total_profit, num_trades
    
    def report_trades(self, trade_details, total_profit, num_trades):
        """
        Print the report of trades.
        """
        print("\n### **Kết quả giao dịch Pair Trading với Dự Đoán Giá Riêng lẻ**\n")
        print(f"- **Số lần giao dịch:** {num_trades} lần.")
        print(f"- **Tổng lợi nhuận:** {total_profit:.2f} VND.\n")
        print("#### **Chi tiết giao dịch:**\n")
        for idx, trade in enumerate(trade_details, 1):
            print(f"{idx}. **Giao dịch:** {trade['strategy']}")
            print(f"   - **Ngày mở vị thế:** {trade['open_date'].date()}")
            print(f"   - **Giá mở FPT:** {trade['open_price_fpt']} VND.")
            print(f"   - **Giá mở CMG:** {trade['open_price_cmg']} VND.")
            
            # Kiểm tra xem 'close_date' có tồn tại không
            close_date = trade.get('close_date', 'N/A')
            close_price_fpt = trade.get('close_price_fpt', 'N/A')
            close_price_cmg = trade.get('close_price_cmg', 'N/A')
            profit = trade.get('profit', 'N/A')
            reason = trade.get('reason', 'N/A')
            
            # Nếu 'close_date' không phải là 'N/A', chuyển sang dạng date
            if close_date != 'N/A' and isinstance(close_date, pd.Timestamp):
                close_date_str = close_date.date()
            else:
                close_date_str = close_date
            
            print(f"   - **Ngày đóng vị thế:** {close_date_str}")
            print(f"   - **Giá đóng FPT:** {close_price_fpt} VND.")
            print(f"   - **Giá đóng CMG:** {close_price_cmg} VND.")
            print(f"   - **Lợi nhuận:** {profit:.2f} VND.")
            print(f"   - **Lý do đóng vị thế:** {reason}\n")
    
    def plot_spread(self):
        """
        Plot the spread and z-score over time.
        """
        plt.figure(figsize=(14, 10))

        # Plot Spread thực tế và Spread dự đoán
        plt.subplot(3, 1, 1)
        plt.plot(self.data['time'], self.data['spread'], label='Spread Thực Tế')
        plt.plot(self.data['time'], self.data['spread_pred'], label='Spread Dự Đoán', alpha=0.7)
        plt.plot(self.data['time'], self.data['mean_spread'], label='Mean Spread')
        plt.fill_between(self.data['time'], 
                         self.data['mean_spread'] + 2*self.data['std_spread'], 
                         self.data['mean_spread'] - 2*self.data['std_spread'], 
                         color='grey', alpha=0.3, label='±2 Std Dev')
        plt.title('Spread Thực Tế và Spread Dự Đoán giữa FPT và CMG')
        plt.legend()

        # Plot Z-Score thực tế
        plt.subplot(3, 1, 2)
        plt.plot(self.data['time'], self.data['z_score'], label='Z-Score Thực Tế')
        plt.axhline(2, color='red', linestyle='--', label='Z > 2')
        plt.axhline(-2, color='green', linestyle='--', label='Z < -2')
        plt.title('Z-Score Thực Tế của Spread')
        plt.legend()

        # Plot Spread Dự Đoán
        plt.subplot(3, 1, 3)
        plt.plot(self.data['time'], self.data['spread_pred'], label='Spread Dự Đoán', color='orange')
        plt.axhline(z_open_threshold, color='red', linestyle='--', label=f'Spread_pred > {z_open_threshold}')
        plt.axhline(-z_open_threshold, color='green', linestyle='--', label=f'Spread_pred < {-z_open_threshold}')
        plt.title('Spread Dự Đoán giữa FPT và CMG')
        plt.legend()

        plt.tight_layout()
        plt.show()
```
